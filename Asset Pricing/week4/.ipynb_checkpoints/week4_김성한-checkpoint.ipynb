## Chapter 6. Mean-variance Portfolio Theory

### 6.1. Asset Return
a. The total return(R) is $ \frac{amount \, received}{amount \, invested} = \frac{X_1}{X_0}$. <br/>
b. The rate of return(r) is $\frac{amount \, received - amount \, invested}{amount\,invested}= \frac{X_1 - X_0}{X_0}$. <br/>
c. $R = 1+r$ and $X_1 = (1+r)X_0$<br/>
d. The process of Short selling, or shorting 
- Borrowing the asset from someone who owns it (such as a brokeage firm).
- Sell the borrowed asset to someone else, receiving an amount $X_0$.
- At a later date, repay your loan by purchasing the asset $X_1$ and return the asset to your lender.
- If the later amount $x_1$ is lower than the original amount, you will have made a profit of $X_0 - X_1$.

$$ R = \frac{-X_1}{-X_0} = \frac{X_1}{X_0}$$

e. Portfolio return: Both the total return and the rate of return of a portfolio of assets are equal to the weighted sum of the corresponding individual asset returns, with the weight of an asset being its relative weight(in purchase cost) in the portfolio.

$$ R = \Sigma_{i=1}^{n}w_iR_i, \, r = \Sigma_{i=1}^{n}w_ir_i, \, \, where \, \Sigma_{i=1}^{n}w_i = 1 $$

### 6.2. Random Variables
a. Random variables: a measurable function from a probability measure space
b. Expected Value
$$ \bar x = E(x) = \Sigma_{i=1}^{m}x_ip_i $$
where $p_i$ is the probability that represents the relative chance of a occurance of $x_i$.

c. Variance
$$ var(y) = E[(y- \bar y)^2] = E[y^2] - E[y]^2 $$
standard devidation: $ \sigma_y = \sqrt{E[(y- \bar y)^2]} $ <br/>
d. Covariance
$$ cov(x_1, x_2) = E[(x_1- \bar {x_1})(x_2- \bar {x_2})] = E[x_1x_2] - E[x_1]E[x_2] $$
- If $\sigma_{12} = 0$ , uncorrelated
- If $\sigma_{12} > 0$ , positively correlated
- If $\sigma_{12} < 0$ , negatively correlated <br/>

e. Correlation coefficient
$$ \rho = \frac{\sigma_{12}}{\sigma_{1} \sigma_{2}}$$ <br/>
f. $var(x+ y)= \sigma_{x}^2 + 2\sigma_{xy} + \sigma_{y}^2 $ <br/>

### 6.4. Portfolio Mean and Variance
a. Mean Return of a Portfolio
Suppose that there are n assets with random rates of return $r_1, r_2, ..., r_n$.These have expected values $E(r_i)= \bar {r_i}$, $i= 1,2, ..., n$. <br/>
- The rate of return of the portfolio in terms of the return of the individual returns is
$$ r = w_1r_1 + w_2r_2 + ... + w_nr_n $$ 
- The expected value is
$$ E(r) = w_1E(r_1) + w_2E(r_2) + ... + w_nE(r_n)$$

b. Variance of Portfolio Return
$$ \sigma^2 = E[(r- \bar r)^2] = E[(\Sigma_{i=1}^{n} w_ir_i- \Sigma_{i=1}^{n} w_i \bar {r_i})^2] = E[(\Sigma_{i,j=1}^{n} w_iw_j(r_i- \bar {r_i})(r_j- \bar {r_j})] = \Sigma_{i,j=1}^{n}w_iw_j\sigma_{ij}$$

c.Diversification <br/>
Portfolios with only a few assets may be subjected to a high degree of risk, represented by a relatively large variance. As general rule, the variance of the return of a portfolio can be reduced by including additional assets in the portfolio, a process referred to as diversification. This process reflects the maxim, "Don't put all your eggs in one basket."

d.Diagram of a Portfolio <br/>

Portfolio diagram lemma : The curve in an $\bar r - \sigma$ diagram defined by nonnegative mixtures of two assets 1 and 2 lies within the triangular region defined by the two original assets and the point on the vertical axis of height $A = \frac{\bar r_1 \sigma_2 + \bar r_2 \sigma_1}{\sigma_1 + \sigma_2}$. <br/>

### 6.5. The Feasible Set
a. feasible set: THe set of points that correspond  to portfolios is called feasible set or feasible region <br/>
b. Two properties in feasible set <br/>
- 1. If there are at least three assets (not perfectly correlated and with different means), the feasible set will be a solid two-demensional region
- 2. The feasible region is convex to the left <br/>

c. The left boundary of a feasible set is called the 'minimum-variance set'. <br/>
d. The minimum-variance set has a characteristic bullet shape. There is a special point on this set having minimum variance. It is termed the minimum-variance point(MVP).

### 6.6 The Markowitz Model
a. The Markowitz Model <br/>
Suppose that there are n assets with random rates of return $r_1, r_2, ..., r_n$.These have expected values $E(r_i)= \bar {r_i}$, $i= 1,2, ..., n$. And the covariances are $\sigma_{ij}$, for $i,j=1,2,...,n$. A portfolio is defined by a set of n weights $w_i, \, i = 1,2,...n$, that sum to 1.(We allow negative weights, corresponding to short selling).To find a minimum- variance portfolio, we fix the mean value at some arbitary value $\bar r$. Then, we find the feasible portfolio of minimum variance that has this mean. <br/>

$$ minimize \, \frac{1}{2} \Sigma_{i,j=1}^{n} w_iw_j\sigma_{ij}$$
$$ subject \, to \,\,\, \Sigma_{i=1}^{n} w_i \bar r_i = \bar r, \,\,\, \Sigma_{i=1}^{n} w_i = 1$$

b. Solution of the Markowitz Problem
Using Lagrange multipliers $\lambda$ and $\mu$, we form the Lagrangian (it is shown specifically at problem 3 in assignment 3).
$$ L =  \frac{1}{2} \Sigma_{i,j=1}^{n} w_iw_j\sigma_{ij} - \lambda(\Sigma_{i=1}^{n} w_i \bar r_i - \bar r) - \mu(\Sigma_{i=1}^{n} w_i - 1)$$

### 6.7. The Two-Fund Theorem
Two efficient funds(portfolios) can be established so that any efficient portfolio can be duplicated, in terms of means variance, as a combination of these two. In other words, all investors seeking efficient portfolios need only invest in combinations of these two funds.

### 6.9. The One-Fund Theorem
There is a single fund F of risky assets such that any efficient protfolio can be constructed as a combination of the fudn F and the risk-free asset.

````python
import numpy as np

def minimize_vol(target_return, er, cov):
    """
    target_ret -> W
    """
    n = er.shape[0]
    init_guess = np.repeat(1/n, n)
    bounds = ((0.0, 1.0),)*n
    return_is_target = {
        'type': 'eq',
        'args': (er,),
        'fun': lambda weights, er: target_return - erk.portfolio_return(weights, er)
    }
    weights_sum_to_1 = {
        'type': 'eq',
        'fun': lambda weights: np.sum(weights) -1 
    }
    results = minimize(erk.portfolio_vol, init_guess,
                       args=(cov,), method="SLSQP",
                       options={'disp':False},
                       constraints=(return_is_target, weights_sum_to_1),
                       bounds=bounds
                      )
    return results.x
````

````python
import pandas as pd
def optimal_weights(n_points, er, cov):
    """
    -> list of weights to run the optimizer on to minimize the vol
    """
    target_rs = np.linspace(er.min(), er.max(), n_points)
    weights = [minimize_vol(target_return, er, cov)for target_return in target_rs]
    return weights
    
def plot_ef(n_points, er, cov):
    """
    Plots the N-assets efficient frontier
    """
    weights = optimal_weights(n_points, er, cov)
    rets = [erk.portfolio_return(w,er) for w in weights]
    vols = [erk.portfolio_vol(w, cov) for w in weights]
    ef = pd.DataFrame({
        "Returns": rets,
        "Volatility": vols
    })
    return ef.plot.line(x="Volatility", y="Returns", style= ".-")
````

````python
from scipy.optimize import minimize
def msr(riskfree_rate, er, cov):
    """
    RiskFree rate + ER + COV -> W
    """
    n = er.shape[0]
    init_guess = np.repeat(1/n, n)
    bounds = ((0.0, 1.0),)*n
    weights_sum_to_1 = {
        'type': 'eq',
        'fun': lambda weights: np.sum(weights) -1 
    }
    def neg_sharpe_ratio(weights, riskfree_rate, er, cov):
        """
        Returns of negative of the sharpe ratio, given weights
        """
        r = erk.portfolio_return(weights, er)
        vol = erk.portfolio_vol(weights, cov)
        return -(r - riskfree_rate)/vol

    results = minimize(neg_sharpe_ratio, init_guess,
                       args=(riskfree_rate, er, cov,), method="SLSQP",
                       options={'disp':False},
                       constraints=(weights_sum_to_1),
                       bounds=bounds
                      )
    return results.x
````

1. Cushion - (Asset Value minus Floor Value)
2. Compute an Allocation to Safe and Risky Assets $\to$ m $\times$ budget
3. Recompute the Asset Value based on the returns

````python
def compound1(r):
    return (1+r).prod()-1

def compound2(r):
    return np.expm1(np.log1p(r).sum())
````

# Back to CPPI

````python
dates = risky_r.index
n_steps = len(dates)
account_value = start
floor_value = start*floor
m = 3
account_history = pd.DataFrame().reindex_like(risky_r)
cushion_history = pd.DataFrame().reindex_like(risky_r)
risky_w_history = pd.DataFrame().reindex_like(risky_r)

for step in range(n_steps):
    cushion = (account_value - floor_value)/account_value
    risky_w = m*cushion
    risky_w = np.minimum(risky_w,1)
    risky_w = np.maximum(risky_w,0)
    safe_w = 1 - risky_w
    risky_alloc = account_value*risky_w
    safe_alloc = account_value*safe_w
    ## update the account value for this time step
    account_value = risky_alloc*(1+risky_r.iloc[step]) + safe_alloc*(1+safe_r.iloc[step])
    # save the value so I can look at the history and plot it etc
    cushion_history.iloc[step] = cushion
    risky_w_history.iloc[step] = risky_w
    account_history.iloc[step] = account_value    
````

### Lab Session-Random Walks and Monte Carlo
# Random Walk Generation

$$ \frac{S_{t+dt}-S_t}{S_t} = \mu dt + \sigma\sqrt{dt}\xi_t$$

````python
import numpy as np

def gbm(n_years=10, n_scenarios=1000, mu=0.07, sigma=0.15, steps_per_year=12, s_0=100.0):
    """
    Evloution of a Stock Price using a Geomertic Brownian Motion Model 
    """
    dt = 1/steps_per_year
    n_steps = int(n_years*steps_per_year)
    xi = np.random.normal(size=(n_steps, n_scenarios))
    rets = mu*dt + sigma*np.sqrt(dt)*xi
    rets = pd.DataFrame(rets)
    # to prices
    prices = s_0*(1+rets).cumprod()
    return prices
````

````python
def gbm0(n_years=10, n_scenarios=1000, mu=0.07, sigma=0.15, steps_per_year=12, s_0=100.0):
    """
    Evloution of a Stock Price using a Geomertic Brownian Motion Model 
    """
    dt = 1/steps_per_year
    n_steps = int(n_years*steps_per_year)
    xi = np.random.normal(size=(n_steps, n_scenarios))
    rets = mu*dt + sigma*np.sqrt(dt)*xi
    rets = pd.DataFrame(rets)
    # to prices
    prices = s_0*(1+rets).cumprod()
    return prices

def gbm1(n_years=10, n_scenarios=1000, mu=0.07, sigma=0.15, steps_per_year=12, s_0=100.0):
    """
    Evloution of a Stock Price using a Geomertic Brownian Motion Model 
    """
    dt = 1/steps_per_year
    n_steps = int(n_years*steps_per_year)
    rets_plus_1 = np.random.normal(loc=(1+mu*dt),scale=(sigma*np.sqrt(dt)),size=(n_steps, n_scenarios))
    # to prices
    prices = s_0*pd.DataFrame(rets_plus_1).cumprod()
    return prices
````

## Problem 2.
There are two risky assets, $A_1,A_2$ such that
- expected returns of A and B:  $Œº_1,Œº_2 $
- standard deviations of A and B: $œÉ_1,œÉ_2 $

Thus, we can construct the portfolio $P=w_1 A+w_2 B$ ($w_1,w_2$ are the weights of risky assets) with these two risky assets.
Using the Lagrange Multiplier Method, find the analytical solution of the efficient frontier.

#### Hint:

When the target expected return is given as $Œº_0$<br/>
Constraints: 1. $w_1 Œº_1+w_2 Œº_2=Œº_0,$ 2. $w_1+w_2=1$<br/>
Step1. Construct Lagrange Function under the given rate $Œº_0$ <br/>
$$ L= \frac{1}{2} ‚àë_{i,j=1}^2 w_i w_j œÉ_{ij} -Œª_1 (‚àë_{i=1}^2w_i Œº_i-Œº_0 )-Œª_2 (‚àë_{i=1}^2w_i-1) $$
Step2.  Find the Optimal weights written by lambda after Partial Differentiation <br/>
$$ \frac{‚àÇL}{‚àÇw_1 }=0,\frac{‚àÇL}{‚àÇw_2 }=0 ‚Üí optimal \, weights \, w^*  \,(function\,of\, Œª_1,\,Œª_2) $$ <br/>
Step3. Apply Optimal weight $w^*$ obtained in the Step2 to constraints, and then find the analytical solution of $w^*$. <br/>
Step4. Derive minimum standard deviation of portfolio in given expected return $Œº_0$ by using the optimal weight $w^*$.

### Solution

- Step 1. Construct Lagrange Function under the given rate $\mu_0$
$$ minimize \, \frac{1}{2} \Sigma_{i,j=1}^{2} w_iw_j\sigma_{ij}$$
$$ subject \, to \,\,\, \Sigma_{i=1}^{2} w_i \bar r_i = \bar r, \,\,\, \Sigma_{i=1}^{2} w_i = 1$$

$$ L= \frac{1}{2} ‚àë_{i,j=1}^2 w_i w_j œÉ_{ij} -Œª_1 (‚àë_{i=1}^2w_i Œº_i-Œº_0 )-Œª_2 (‚àë_{i=1}^2w_i-1) $$
$$ \to L= \frac{1}{2}({w_1}^2{œÉ_{1}}^2 +2w_1w_2œÉ_{12}+{w_2}^2{œÉ_{2}}^2) -Œª_1 (w_1 Œº_1 +w_2 Œº_2 -Œº_0 )-Œª_2 (w_1+w_2 -1) $$
$where \,\, \sigma_{12} = \sigma_{21}$.

- Step2. Find the Optimal weights written by lambda after Partial Differentiation <br/>
$ \frac{‚àÇL}{‚àÇw_1 }= \frac{1}{2}(2w_1{œÉ_{1}}^2 + 2w_2œÉ_{12})-Œª_1Œº_1 - Œª_2=0 $  <br/>
$ \frac{‚àÇL}{‚àÇw_2 }= \frac{1}{2}(2w_2{œÉ_{2}}^2 + 2w_1œÉ_{12})-Œª_1Œº_2 - Œª_2=0 $  <br/>

Rewrite the above equations, <br/>
$$ w_1{œÉ_{1}}^2 + w_2œÉ_{12}-Œª_1Œº_1 - Œª_2=0 \,\,\,\,(A)$$
$$ w_2{œÉ_{2}}^2 + w_1œÉ_{12}-Œª_1Œº_2 - Œª_2=0 \,\,\,\,(B)$$

Using (A) and (B), we can derive optimal $w^*$.

$$ {w_1}^* = \frac{(\mu_1{\sigma_2}^2 - \mu_2 \sigma_{12})\lambda_1 + ({\sigma_2}^2 - \sigma_{12})\lambda_2}{{\sigma_1}^2{\sigma_2}^2-{\sigma_{12}}^2}$$
$$ {w_2}^* = \frac{(\mu_2 {\sigma_1}^2 - \mu_1\sigma_{12})\lambda_1 + ({\sigma_1}^2 - \sigma_{12})\lambda_2}{{\sigma_1}^2{\sigma_2}^2-{\sigma_{12}}^2}$$

- Step3. Apply Optimal weight $w^*$ obtained in the Step2 to constraints, and then find the analytical solution of $w^*$. <br/>

Let $A=(\mu_1{\sigma_2}^2 - \mu_2 \sigma_{12})$, $B={\sigma_2}^2 - \sigma_{12}$, $C=\mu_2 {\sigma_1}^2 - \mu_1\sigma_{12}$, $D={\sigma_1}^2 - \sigma_{12}$, $E={\sigma_1}^2{\sigma_2}^2-{\sigma_{12}}^2$.

Rewrite the constraints uisng A,B,C,D and E,

$ w_1 + w_2 = 1 \to (A+C)\lambda_1 + (B+D)\lambda_2 = E $ <br/>
$ w_1\mu_1 + w_2\mu_2 = \mu_0 \to (A\mu_1+C\mu_2)\lambda_1 + (B\mu_1+D\mu_2)\lambda_2= E\mu_0 $

Solving for $\lambda_1$ and $\lambda_2$,
$${\lambda_1}^* = \frac{E[B(\mu_1 - \mu_0)+D(\mu_2 - \mu_0)]}{(AD-BC)(\mu_2-\mu_1)} $$
$${\lambda_2}^* = \frac{-E[A(\mu_1 - \mu_0)+C(\mu_2 - \mu_0)]}{(AD-BC)(\mu_2-\mu_1)}$$

Since ${w_1}^*$ and ${w_2}^*$ is
$$ {w_1}^* = \frac{A\lambda_1 + B\lambda_2}{E} {w_2}^* = \frac{C\lambda_1 + D\lambda_2}{E} $$

Substituting ${\lambda_1}$, ${\lambda_2}$ with ${\lambda_1}^*$, ${\lambda_1}^*$,

$$ {w_1}^* = \frac{A{\lambda_1}^* + B{\lambda_2}^*}{E} = \frac{\mu_2 - \mu_0}{\mu_2 - \mu_1}$$
$$ {w_2}^* = \frac{C{\lambda_1}^* + D{\lambda_2}^*}{E} = \frac{\mu_0 - \mu_1}{\mu_2 - \mu_1}$$


- Step4. Derive minimum standard deviation of portfolio in given expected return $Œº_0$ by using the optimal weight $w^*$.

The standard deviation at the solution is $\sqrt{{w_1}^2{œÉ_{1}}^2 +2w_1w_2œÉ_{12}+{w_2}^2{œÉ_{2}}^2}$.
$$ \sigma_{min} = \frac{\sqrt{(\mu_2 - \mu_0)^2{\sigma_1}^2 +2(\mu_2 - \mu_0)(\mu_0 - \mu_1)\sigma_{12}+(\mu_0 - \mu_1)^2{\sigma_2}^2}}{\mu_2 - \mu_1}$$

## Problem 3.
Jin‚Äôs friend is planning to invest \\$1 million in a rock concert to be held 1 year from now. The friend figures that he will obtain \\$3 million revenue from his \\$1 million investment-unless, my goodness, it rains. If it rains, he will lose his entire investment. There is a 50% chance that it will rain the day of the concert. Gain suggests that he buy rain insurance. He can buy one unit of insurance for \\$0.5. and this unit pays \\$1 if it rains and nothing if it does not. He may purchase as many as he wishes, up to \\$3 million.<br/>
a) What is the expected rate of return on his investment if he buys $u$ units of insurance? (The cost of insurance is in addition to his \\$1 million investment.)

### Solution
Expected rate of return = $\frac{amount\,received - amount\,invested}{amount\,invested}$

Expected amount received = $0.5(3,000,000) + 0.5u = 1,500,000 + 0.5u $ <br/>
Amount invested = $1,000,000 + 0.5u$<br/>
$E(r) = \frac{1,500,000 + 0.5u - 1,000,000 - 0.5u}{1,000,000 + 0.5u}$ <br/>
$\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,= \frac{500,000}{1,000,000 + 0.5u}$

b) What number of units will minimize the variance of his return? What is this minimum value? And what is the corresponding expected rate of return? <br/>
Hint: Before calculating a general expression for variance, think about a simple answer.

### Solution
To make minimize variance, make a same amount of returns at any situations. That makes the vairance zero.

If there is no rain, the amount of return is $3,000,000 - 1,000,000 - 0.5u$.
If there is rain, the amount of return is $0 - 1,000,000 - 0.5u + u$.

$3,000,000 - 1,000,000 - 0.5u = 0 - 1,000,000 - 0.5u + u$
$\therefore u = 3,000,000$

- the minimum variance = 0
- this minimum value = \\$500,000 ($\because$ buy 3 million units of insurance)
- expected rate of return = $\frac{500,000}{1,000,000 + 0.5u} = \frac{500,000}{2,500,000} = 20\% $
