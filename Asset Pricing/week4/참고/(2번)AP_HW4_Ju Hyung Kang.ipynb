# Chapter 6. Mean-Variance Portfolio Theory

## 6.1 Asset Return



Total return: $R=\frac{X_1}{X_0}$

Rate of return: $r=\frac{X_1-X_0}{X_0}$

$$
R=1+r
$$

#### Portfolio Return

$$
R=\sum^n_{i=1}w_iR_i
$$

$$
r=\sum^n_{i=1}w_ir_i
$$

## 6.2 Random Variables

Random variables

Probability density function

Expected Value: $E(X)=\sum_{i}x_ip_i$.

Variance: $Var(X)=E[(X-E(X))^2]$.

Standard deviation: $\sigma_X=\sqrt{Var(X)}$.

Covariance: $Cov(X_1,X_2)=E[(X_1-E(X_1))(X_2-E(X_2))]$

Correlation coefficient: $\rho_{12}=\frac{\sigma_{12}}{\sigma_1\sigma_2}$

## 6.4 Portfolio Mean and Variance

Portfolio Mean: $E(r)=w_1E(r_1)+w_2E(r_2)+\cdots+w_nE(r_n)$

Portfolio Variance: $\sigma^2=\sum^n_{i,j=1}w_iw_j\sigma_{ij}$



Portfolio diagram lemma: The curv in an $\overline{r}-\sigma$ diagrom defined by nonnegative mixtures of two assets 1 and 2 lies within the triangular region defined by the two original assets and the point on the vertical axis of height $A=(\overline{r}_1\sigma_2+\overline{r}_2\sigma_1)/(\sigma_1+\sigma_2)$.

## 6.5 The Feasible Set

Feasible set(Feasible region): The set of points that correspond to portfolios

Properties
1. IF there are at least three assets (not perfectly correlated and with different means), the feasible set will be a solid two-dimensional region.
2. The feasible region is convex to the left.

#### The Minimum-Variance Set and the Efficient Frontier

The minimum-variance set: The left boundary of a feasible set

The Mininum-variance point(MVP)

Efficient frontier: the upper portion of the minimum-variance set

## 6.6 The Markowitz Model

Fix the mean at $\overline{r}$.

Find the feasible portfolio of minimum variance that has this mean.

#### Solution of the Markowitz Problem

We can find the conditions for a solution to this problem using Lagrangemultipliers $\lambda$ and $\mu$.

1. Form the Lagrangian

$$
L=\frac{1}{2}\sum^n_{i,j=1}w_iw_j\sigma_{ij}-\lambda\left(\sum^n_{i=1}w_i\overline{r}_i-\overline{r}\right)-\mu\left(\sum^n_{i=1}w_i-1\right)
$$

For two variables,

$$
L=\frac{1}{2}(w_1^2\sigma_1^2+w_1w_2\sigma_{12}+ w_2w_1\sigma_{21}+w_2^2\sigma^2_2)-\lambda(\overline{r}_1w_1+\overline{r}_2w_2-\overline{r})-\mu(w_1+w_2-1).
$$

2. Differentiate the Lagrangian with respect to each variable $w_i$ and set this derivative to zero.

$$
\frac{\partial L}{\partial w_1} = \frac{1}{2}(2\sigma^2_1w_1+\sigma_{12}w_2+\sigma_{21}w_2)-\lambda \overline{r}_1-\mu=0
$$

$$
\frac{\partial L}{\partial w_2} = \frac{1}{2}(\sigma_{12}w_1+\sigma_{21}w_1+2\sigma^2_2w_2)-\lambda \overline{r}_2-\mu=0
$$

=>
$$
\sigma^2_1w_1+\sigma_{12}w_2-\lambda\overline{r}_1-\mu=0
$$
$$
\sigma_{12}w_1+\sigma^2_2w_2\lambda\overline{r}_2-\mu=0
$$

3. Sove for $w_1, w_2, \lambda, \mu$.

## 6.7 The Two-Fund Theorem

The two-fund theorem: Two efficient funds(portfolios) can be established so that any efficient portfolio can be duplicated, in terms of mean and variance, as a combination of these two. In other words, all investors seeking efficient portfolios need only invest in combinations of these two funds.

## 6.8 Inclusion of a Risk-Free Asset

$\alpha$ for the risk-free asset and $1-\alpha$ for the risky asset

SD: $(1-\alpha)\sigma$

## 6.9 The One-Fund Theorem

The one-fund theorem: There is a single fund F of risky assets such that any efficient portfolio can be constructed as a combination of the fund F and the risk-free asset.

````python
def minimize_vol(target_return, er, cov):
    """
    target_return -> W
    """
    n = er.shape[0]
    init_guess = np.repeat(1/n,n)
    bounds = ((0.0, 1.0),)*n
    return_is_target = {
        'type': 'eq',
        'args': (er, ),
        'fun': lambda weights, er: target_return - erk.portfolio_return(weights, er)
    }
    weights_sum_to_1 = {
        'type': 'eq',
        'fun': lambda weights: np.sum(weights)-1
    }
    results = minimize(erk.portfolio_vol, init_guess,
                      args=(cov, ), method="SLSQP",
                      options={'disp': False},
                      constraints = (return_is_target, weights_sum_to_1),
                      bounds=bounds
                      )
    return results.x
````

````python
def optimal_weights(n_points, er, cov):
    """
    -> List of weights to run the optimizer on to minimize the vol
    """
    target_rs = np.linspace(er.min(), er.max(), n_points)
    weights = [minimize_vol(target_return, er, cov) for target_return in target_rs]
    return weights
    
def plot_ef(n_points, er, cov):
    """
    Plots the multi-asset efficient frontier
    """
    weights = optimal_weights(n_points, er, cov)
    rets = [portfolio_return(w, er) for w in weights]
    vols = [portfolio_vol(w, cov) for w in weights]
    ef = pd.DataFrame({
        "Returns": rets,
        "Volatility": vols
    })
    return ef.plot.line(x="Volatility", y="Returns", style='.-')
    
````

## Lab Session-Locating the Max Sharpe Ratio Portfolio

````python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import edhec_risk_kit as erk

ind = erk.get_ind_returns()
er = erk.annualize_rets(ind["1996":"2000"], 12)
cov = ind["1996":"2000"].cov()
````

````python
def msr(riskfree_rate, er, cov):
    """
    Risk Free Rate + ER + COV -> W
    """
    n = er.shape[0]
    init_guess = np.repeat(1/n,n)
    bounds = ((0.0, 1.0),)*n
    weights_sum_to_1 = {
        'type': 'eq',
        'fun': lambda weights: np.sum(weights)-1
    }
    
    def neg_sharpe_ratio(weights, riskfree_rate, er, cov):
        """
        Returns the negative of the sharpe ratio, given weights
        """
        r = erk.portfolio_return(weights, er)
        vol = erk.portfolio_vol(weights, cov)
        return -(r-riskfree_rate)/vol

    results = minimize(neg_sharpe_ratio, init_guess,
                      args=(riskfree_rate, er, cov, ), method="SLSQP",
                      options={'disp': False},
                      constraints = (weights_sum_to_1),
                      bounds=bounds
                      )
    return results.x
````

````python
ax = erk.plot_ef(20, er, cov)
ax.set_xlim(left=0)
rf = 0.1
w_msr = msr(rf, er, cov)
r_msr = erk.portfolio_return(w_msr,er)
vol_msr = erk.portfolio_vol(w_msr,cov)
# Add CML
cml_x = [0, vol_msr]
cml_y = [rf, r_msr]
ax.plot(cml_x, cml_y, color="green", marker="o", linestyle="dashed")
````

## Problem 2

There are two risky assets, $A_1,A_2$, such that

expected returns of A and B:  $\mu_1,\mu_2$
standard deviations of A and B: $\sigma_1,\sigma_2$

Thus, we can construct the portfolio $P=w_1 A+w_2 B$ ($w_1,w_2$ are the weights of risky assets) with these two risky assets.

Using the Lagrange Multiplier Method, find the analytical solution of the efficient frontier

Hint:

When the target expected return is given as $\mu_0$

Constraints:

1. $w_1 \mu_1+w_2 \mu_2=\mu_0$
2. $w_1+w_2=1$

Step1. Lagrange Function under the given rate $\mu_0$

$$
L=\frac{1}{2}\sum^2_{i,j=1}w_iw_j\sigma_{ij}-\lambda_1\left(\sum^2_{i=1}w_i\mu_i-\mu_0\right)-\lambda_2\left(\sum^2_{i=1}w_i-1\right)
$$

Step2. Find the Optimal weights written by lambda after Partial Differentiation.

$$
\frac{\partial L}{\partial w_1}=0, \frac{\partial L}{\partial w_2}=0 \rightarrow \text{optimal weights} \,\,w^*(\text{function of}\,\, \lambda_1,\lambda_2).
$$

Step3. Apply Optimal weight $w^*$ obtained in the Step2 to constraint formula, and then find the analytical solution of $w^*$.

Step4. Derive minimum standard deviation of portfolio in given expected return $\mu_0$ by using the optimal weight $w^*$.

Solution)

Step1:

$$
L=\frac{1}{2}(w_1^2\sigma_1^2+2w_1w_2\sigma_{12}+w_2\sigma^2_2)-\lambda_1(\mu_1w_1+\mu_2w_2-\mu_0)-(w_1+w_2-1)
$$

Step2:

$$
\frac{\partial L}{\partial w_1} = w_1^*\sigma^2_1 + w_2^* \sigma_{12} - \lambda_1 \mu_1 - \lambda_2=0
$$
$$
\frac{\partial L}{\partial w_2} = w_1^*\sigma_{12} + w_2^* \sigma^2_{2} - \lambda_1 \mu_2 - \lambda_2=0
$$

Step3:

Since $w_1+w_2=1$,
$$
w_1^* = \frac{\lambda_1 \mu_1+\lambda_2-\sigma_{12}}{\sigma^2_1-\sigma_{12}}=\frac{\lambda_1\mu_2+\lambda_2-\sigma^2_2}{\sigma_{12}-\sigma^2_2}
$$
and
$$
w_2^* = \frac{\lambda_1 \mu_1+\lambda_2-\sigma^2_{1}}{\sigma_{12}-\sigma^2_1}=\frac{\lambda_1\mu_2+\lambda_2-\sigma_{12}}{\sigma^2_2-\sigma_{12}}
$$

Step4:


The minimum standard deviation of portfolio in given expected return $\mu_0$ by using the optimal weight $w^*$ is

$$
\sigma_{min} = \sqrt{\sigma^2_{min}} = \sqrt{(w_1^*)^2\sigma_{1}^2+2(w_1^*)(w_2^*)\sigma_{12}+(w_2^*)^2\sigma_{2}^2}
$$

## Problem 3

Jin’s friend is planning to invest \\$1 million in a rock concert to be held 1 year from now. The friend figures that he will obtain \\$3 million revenue from his \\$1 million investment-unless, my goodness, it rains. If it rains, he will lose his entire investment. There is a 50\% chance that it will rain the day of the concert. Gain suggests that he buy rain insurance. He can buy one unit of insurance for \\$0.5. and this unit pays $1 if it rains and nothing if it does not. He may purchase as many as he wishes, up to \\$3 million.


a) What is the expected rate of return on his investment if he buys $u$ units of insurance? (The cost of insurance is in addition of his \\$1 million investment.)

Solution) Define a random variable, $X$, for the concert revenue and a random variable, $I$, for the outcome of insurance. Then,

$$
\begin{align*}
E(r) &= E\left(\frac{X+uI-(1000000+0.5u)}{1000000+0.5u}\right)\\
&=\frac{E(X+uI)-(1000000+0.5u)}{1000000+0.5u}\\
&=\frac{E(X)+uE(I)-(1000000+0.5u)}{1000000+0.5u}\\
&=\frac{0.5 \cdot 3000000 + 0.5 \cdot 0 +u(0.5 \cdot 0 + 0.5 \cdot 1)-(1000000+0.5u)}{1000000+0.5u}\\
&=\frac{1500000 + 0.5u-1000000-0.5u}{1000000+0.5u}\\
&=\frac{500000}{1000000+0.5u}\\
\end{align*}
$$

b) What number of units will minimize the variance of his return? What is this minimum value? And what is the corresponding expected rate of return?

Hint: Before calculating a general expression for variance, think about a simple answer.

Solution) Let $K=1000000+0.5u$, then

$$
\begin{align*}
Var(r) &= Var\left(\frac{X+uI-(1000000+0.5u)}{1000000+0.5u}\right)\\
&=\frac{Var(X+uI)-K}{K}\\
&=\frac{1}{K^2}(Var(I)u^2+ 2Cov(X,I)u+Var(X))\\
&=\frac{1}{K^2}Var(I)\left(u^2+ 2\frac{Cov(X,I)}{Var(I)}u+\frac{Var(X)}{Var(I)}\right)\\
\end{align*}
$$

Thus, $Var(r)$ will be minimized with $u=-\frac{Cov(X,I)}{Var(I)}$.

Note that

$$
\begin{align*}
-\frac{Cov(X,I)}{Var(I)}&=-\frac{\sum(X-E(X))(I-E(I))}{\sum(I-E(I))^2}\\
&=-\frac{(3000000-1500000)(0-0.5) + (0-1500000)(1-0.5)}{(0-0.5)^2+(1-0.5)^2}\\
&=3000000
\end{align*}
$$

Thus, the variance is minimized by purchasing 3 million units of insurance.
