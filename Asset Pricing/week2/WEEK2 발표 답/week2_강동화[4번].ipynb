### Chapter5. Risk, Return, and the Historical Record
1. The Nominal rate of interest is the equilibrium real interest rate plus the expected rate of inflation.
In general, we can directly observe only nominal interest rates; from them, we must infer expected real rates, using inflation forecasts. 
Assets with guaranteed nominal rates are risky in real terms because the future inflation rate is uncertain.
 

2. The equilibrium expected rate of return on any security is the sum of the equilibrium real rate of interest,
the expected rate of inflation, and a security-specific risk premium

 

3. Investors face a trad-off between risk and expected return.
Historical data confirm our intuition that assets with low degrees of risk should provide lower returns on average
than do those of higher risk.

4. Historical returns on stocks exhibit somewhat more frequent large negative deviation from the mean than
would be predicted from normal distribution. 
The lower partial standard deviation, skew, and kurtosis of the actual distribution quantify the deviation from normality 

5. Widely used measures of tail risk are value at risk(VaR), and expected shorfall of, equivalently, conditional tail expectations.
VaR measures the loss that will be exceed with a speciried probality.
Expected shortfall(ES) measures the expected rate of return conditional on the portfolio falling below a certain value.

6. Investments in risky portfolios do not become safer in the long run.
The longer a risky investment is held, the greater the risk.
The basis of the argument that stocks are safe in the long run is the fact that the probability of an investment shortgfall bacomes smaller.




### Chaper6. Capital Allocation to Risky Assets
1. Investor's preferences toward the expected return and volatility of portfolio may be expressed by utility funtion that is higher for higher expected returns and lower for higher portfolio variances.
Indifference curves show that more risk-averse investors will apply geater penalties for risk.
 

2. An investor's risky portfolio can be characterized by its reward-to-volatility or Sharpe-ratio(S = [E(rP) − rf]/σP ).
 Sharpe ratio is also the slope of the CAL, the line that, when graphed, goes from the risk-free asset through the risky asset.
An Investor would prefer a steeper-sloping CAL, because that means higher expected return for any level of risk.
 

3. The optimal position, y*, in the risky asset, is proportional to the risk premium and inversely proportional to the variance and degree of riskt aversion.
 Graphically, this portfolio represnets the potint at which the indifference curve is tangent to the CAL.
