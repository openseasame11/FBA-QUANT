# 1. Summary

### Chap7. THE CAPITAL ASSET PRICING MODEL  

#### 7.1 Market Equilibrium
1. 모두가 유일 펀드만 구입하고 이 펀드가 시장을 형성하게 되면, 유일 펀드가 시장 포트폴리오가 된다. 이 때 각 자산의 비중은 시장 전체의 자본 비중과 같다.
2. 모든 투자자가 동일한 추정 모수로 평균-분산 포트폴리오를 주문하면, 모든 투자자들은 수요-공급의 변화에 따라 최적 포트폴리오를 곧바로 재검토한다. 이러한 과정을 수요와 공급이 일치할 때까지 계속하는데, 이를 균형 이론(equilibrium)이라고 한다.

#### 7.2 The Capital Market Line
1. $ \bar r - \sigma $ 도표에 시장 포트폴리오 M을 label하면 효율적인 집합으로 구성되는 직선을 그릴 수 있는데, 이를 자본시장선(CML)이라고 한다. 

$$ \bar r = r_f + \frac {\bar r_M - r_f}{\sigma_M} \sigma $$ 

2. 자본 시장의 기울기 $K = (\bar r_M - r_f) / \sigma_M$는 위험가격이다. 

#### 7.3 The Pricing Model
1. 자본자산 가격결정 모형(Capital Asset Pricing Model; CAPM) : 만약 시장 포트폴리오 $M$이 효율적이라면, 자산 $i$의 기대수익률 $\bar r_i$는 다음을 만족한다.

$$\bar r_i - r_f = \beta_i(\bar r_M - r_f)$$

$$\beta_i = \frac{\sigma_{iM}}{\sigma^2_M}$$

2. CAPM은 자산의 초과 기대수익률이 시장 포트폴리오의 초과 기대수익률에 비례상수 $\beta$를 곱하여 표현된다. 이 때 $\beta$는 자산과 시장포트폴리오의 공분산을 시장 포트폴리오로 정규화한 형태로 해석할 수 있기 때문에 CAPM 공식을 통해 자산의 초과 기대수익률이 시장 포트폴리오와의 공분산에 비례한다는 것을 알 수 있다.
3. 포트폴리오의 베타는 개별 자산의 베타의 가중 평균으로 나타내진다.

#### 7.4 The Security Market Line
1. 앞서 설명한대로 CAPM의 선형 관계를 이용하면 다음과 같은 두가지 형태의 증권시장선(security market line)을 나타낼 수 있다.

##### Systematic Risk
2. CAPM에서 베타는 중요한 위험 척도이며 자산 $i$의 수익률을 $r_i = r_f + \beta_i(r_M - r_f) + \epsilon_i $와 같이 표현한다.
3. 앞선 식에 기대값을 취하면 CAPM에 의해 $E(\epsilon_i)=0$이 되고 베타에 대한 정의를 이용하면 $cov(\epsilon_i, \sigma_M)=0$을 구할 수 있다. 따라서 $\sigma_i^2 = \beta_i^2\sigma_M^2 + var(\epsilon_i)$
4. 이 때, 우변의 앞 부분 $\beta_i^2\sigma_M^2$은 전체 시장과 관련된 위험, 즉 체계적 위험(systematic risk)을 나타낸다. 다시 말해 베타가 0이 아닌 이상 모든 자산은 체계적 위험을 포함한다.
5. 뒷 부분 $var(\epsilon_i)$은 비체계적(nonsystematic), 특별한(special) 위험을 나타낸다. 이는 시장과 상관관계가 없고 분산투자로 줄일 수 있다.

#### 7.7 CAPM as a Pricing Formula
1. 자산 매수 가격을 $P$, 매도 가격을 $Q$라고 정의하면 수익률 $r = (Q-P) / P$이다. 이를 CAPM 공식에 넣으면 $\frac{\bar Q - P}{P} = r_f + \beta(\bar r_M - r_f)$이고, $P$에 대해서 식을 전개하면 $$ P = \frac{\bar Q}{1 + r_f  + \beta (\bar r_M - r_f)}$$
##### Linearity of Pricing and the Certainty Equivalent Form
2. $r = Q/P - 1$을 이용하여, 베타를 구하면 $\beta = \frac {cov[(Q/P - 1), r_M]}{\sigma_M^2}$이고, 위 식을 정리하면 $$\beta = \frac {cov(Q,r_M)}{P\sigma_M^2}$$
위 공식을 가격 결정 공식에 대입하고, P로 나누면 $$1 = \frac{\bar Q}{P(1+r_f) + cov(Q,r_M)(\bar r_M - r_f)/\sigma_M^2}$$
마지막으로 위 식을 P에 관해 정리하면 $$ P = \frac {1}{1 + r_f}[\bar Q - \frac{cov(Q,r_M)(\bar r_M - r_f)}{\sigma_M^2}]$$
3. 대괄호 안의 식을 Q의 확정등가라고 한다. 대괄호 안의 수식이 모두 Q에 대해 선형이므로 확정등가 가격 결정 모형은 가격 결정 공식이 선형임을 보여준다.

#### 8.2 Factor Models
##### Single-Factor Model
$$ r_i = a_i + b_if + e_i,\space\space i = 1,2,...,n \space\space ... \space\space (8.1)$$ 

$a_i와 \space b_i$는 고정된 상수, $e_i$는 오차를 나타내는 확률변수, 오차와 요인 f는 관계가 없다고 가정.

1. $a_i$는 $i$에 대한 직선의 y절편, $b_i$는 요인에 대한 비율, 즉 요인에 대한 수익률 민감도
2. Single-Factor Model의 평균-분산 모수
$$ \bar r_i = a_i + b_i\bar f + e_i $$

$$ \sigma_i^2 = b_i^2\sigma_f^2 + \sigma_{e_i}^2 $$

$$ \sigma_{ij} = b_ib_j\sigma_f^2,\space i≠j $$

$$ b_i = cov(r_i, f)\space / \space\sigma_f^2 $$
##### Portfolio Parameters
$$ \sum^n_{i=1} w_i = 1 \space일\space때, \space\space r = \sum^n_{i=1} w_ia_i + \sum^n_{i=1} w_ib_if + \sum^n_{i=1} w_ie_i $$

$$ ⇒ r = a+bf+e $$

$$ \sigma_e^2 = E[(\sum^n_{i=1} w_ie_i)(\sum^n_{j=1} w_je_j)] = E(\sum^n_{i=1} w_i^2e_i^2) = \sum^n_{i=1} w_i^2\sigma_{e_i}^2 \space\space \because e_i는 \space 서로 \space상관관계가 \space 없다고 \space 가정$$

3. 다시 식 (8.1)을 보면 $e_i$에서 발생하는 위험은 분산투자로 제거할 수 있으므로 분산가능 위험, $b_if$에서 발생하는 위험은 분산투자로 제거 불가능하므로 분산불가능 혹은 체계적 위험이라 한다.
##### Multifactor Models
$$r_i =a_i+ b_{1i}f_1 + b_{2i}f_2 + e_i$$

오차에 대한 기댓값은 0, 오차와 두 요인간의 상관관계가 없고, 각 자산 오차간의 상관관계도 없다고 가정. 두 요인 간 상관관계는 존재



#### 8.3 The CAPM as a Factor Model 
##### CAPM은 Single-Factor Model의 특수한 경우로 표현(유도)할 수 있다.
시장수익률 $r_M$을 요인으로 사용하는 Single-Factor Model에서 편의상 상수 $r_f$를 시장수익률 요인과 개별 주식 수익률에 대해 빼면 Factor Model은 다음과 같다.

$$ r_i - r_f = \alpha_i + \beta_i(r_M - r_f) + e_i \space\space ... \space\space (8.2)$$

1. 식 8.2에서 $e_i = 0$으로 가정하면 증권 특성방정식(characteristic equation)을 얻을 수 있고 아래와 같이 증권특성선(characteristic line)을 그릴 수 있다.



2. 이 방정식에 대한 기댓값은 $\bar r_i - r_f = \alpha_i + \beta_i(\bar r_M - r_f)$이고, 이 식은 $\alpha_i$를 제외하면 CAPM과 같다. 즉 CAPM은 이 방정식에서 $\alpha_i = 0$을 가정한 식이다.

3. 다시 식 8.2의 양변에 $r_M$에 대한 공분산을 취하여 얻은 $\beta_i = \frac {\sigma_{iM}}{\sigma^2_M}$와 같고, 이는 CAPM에서의 베타에 대한 수식과 일치한다.

#### 8.4 Arbitrage Pricing Theory
##### Simple Version of APT
1. 두 자산 $i,j$에 대해 $ r_i = a_i + b_if, \space\space r_j = a_j + b_jf$을 고려하자. $b_i≠b_j$을 만족할 때, $w_i = w, \space w_j = 1 - w$ 라고 하면 포트폴리오의 수익률은 다음과 같다.

$$ r = wa_i + (1-w)a_j + [wb_i + (1-w)b_j]f$$ 

이 때, $w = b_j \space/\space (b_j - b_i)$라고 하고, 위 식에 대입하면 다음과 같은 수익률을 얻는다.

$$ r = wa_i + (1-w)a_j = \frac {a_ib_j}{b_j - b_i} + \frac {a_jb_i}{b_i - b_j}$$

위의 수익률을 $\lambda_0$라고 하고 무위험 자산이 존재하는 경우, $\lambda_0 = r_f$라고 하면

$$ r = wa_i + (1-w)a_j = \frac {a_ib_j}{b_j - b_i} + \frac {a_jb_i}{b_i - b_j} = \lambda_0$$

$$ \lambda_0(b_j - b_i) = a_ib_j - a_jb_i$$

$$ ⇒ \frac {a_j - \lambda_0}{b_j} = frac {a_i - \lambda_0}{b_i} $$

$$ 모든 \space i,j에 \space 대해 \space 만족해야\space하므로 \space⇒  \frac {a_i - \lambda_0}{b_i} = c \space ⇒ \space a_i = \lambda_0 + b_ic$$

위의 정보를 이용하여 기대 수익률을 구하는 공식을 유도하면 다음과 같다.

$$ \bar r_i = a_i + b_i\bar f = \lambda_0 + b_ic + b_i\bar f = \lambda_0 + b_i\lambda_1 \space\space (\lambda_1 = c + \bar f) \space\space ... \space\space (8.3)$$

2. 식 8.3에서 요인 $f = r_M$이라 하면, $\lambda_0 = r_f, \space\lambda_1 = \bar r_M - r_f, \space b_i = \beta_i$로 설정할 수 있다. 이 때, 차익거래 가격 결정 이론 공식 (8.3)은 CAPM 공식과 같아진다. 
##### Well-Diversified Portfolios
3. n개의 자산이 존재하고 자산 i에 대한 수익률이 다음을 만족한다고 하자.

$$ r_i = a_i + \sum^m_{j=1} b_{ij}f_j + e_i, \space\space E(e_i) = 0, \space\space E(e_i^2) = \sigma_{e_i}^2 $$

$$ ⇒ \sum^m_{i=1} w_i = 1 \space 일 \space때, \space 포트폴리오의 \space 수익률\space r = a + \sum^m_{j=1} b_jf_j + e $$

이때, $W \approx 1$에 대해 $w_i ≤ W/n$을 만족하면 잘 분산된 포트폴리오(well diversified portfolio)라고 정의한다.

# 2. Make a Problem

##### Q. 수익률이 $\alpha r_f + (1-\alpha)r_M$인 포트폴리오에 대해, 확정등가 가격 결정 모형 공식을 이용해 $100 펀드 자산의 가치가 100달러임을 보이시오.  

##### A.

수익률 $r = \alpha r_f + (1-\alpha)r_M$이고, $Q = P(1+r) = P(1 + \alpha r_f + (1-\alpha)r_M)$와 같이 주어진다.

$$ ⇒ cov(Q, r_M) = cov(P(1 + \alpha r_f + (1-\alpha)r_M), r_M) = P(1-\alpha)\sigma_M^2$$

위 과정으로 도출된 $cov(Q, r_M) = P(1-\alpha)\sigma_M^2$을 확정등가 가격 결정 모형 공식 $P = \frac {1}{1 + r_f}[\bar Q - \frac{cov(Q,r_M)(\bar r_M - r_f)}{\sigma_M^2}]$에 대입하면

$$ P = \frac {1}{1+r_f}[P(1 + \alpha r_f + (1-\alpha)r_M) - \frac {P(1-\alpha)\sigma_M^2(r_M - r_f)}{\sigma^2_M}] $$

$$ = \frac {1}{1+r_f}[P(1 + \alpha r_f + (1-\alpha)r_f)] = \frac {1}{1+r_f}[P(1+r_f)] = P$$

따라서 100달러 펀드자산의 가치는 100달러이다.

### 1.

#### constraints

$$ w^T\mu = \mu_0 \space\space ... \space\space (1)$$

$$ w^T1_N = 1 \space\space ... \space\space (2)$$

위 두 제약식이 주어져 있고, 포트폴리오의 기대수익이 $\mu_0$로 주어져 있을 때 포트폴리오의 분산($\sigma^2_p = w^TΣw$)을 최소화하는 $w$를 찾아야 한다.

#### Lagrange Function under given rate $\mu_0$
$ \delta_1, \delta_2 \in R $에 대해 라그랑주 함수를 작성하면 다음과 같다.

$$ L(w,\delta_1, \delta_2) = w^TΣw + \delta_1(\mu_0 - w^T\mu) + \delta_2(1 - w^T1_N) $$

#### Find $w^*$
최종 목표값인 포트폴리오 최적화 가중치를 $w^*$라고 하자. 이를 구하기 위해 라그랑주 함수를 $w$에 대해 편미분하고 이 값이 0과 같다고 놓으면 다음과 같다.

$$ \frac {\partial L(w^*,\delta_1, \delta_2)}{\partial w} = 2Σw^* - \delta_1\mu - \delta_2 1_N = 0 $$

이 식을 $w^*$에 대해 정리하면 다음과 같다.

$$ w^* = Σ^{-1} (\frac{\delta_1}{2}\mu + \frac{\delta_2}{2}1_N) $$

앞서 도출한 $w^*$를 (1),(2) 제약식에 대입하면 다음과 같다.

$$
\begin {cases}
\mu_0 = \frac{\delta_1}{2}\mu^TΣ^{-1}\mu + \frac{\delta_2}{2}\mu^TΣ^{-1}1_N = \frac{\delta_1}{2}B + \frac{\delta_2}{2}A\\
1 = \frac{\delta_1}{2}1_NΣ^{-1}\mu + \frac{\delta_2}{2}1_N^TΣ^{-1}1_N = \frac{\delta_1}{2}A + \frac{\delta_2}{2}C\\
\end {cases}
$$

$$ ⇒ \begin{bmatrix} \mu_0 \\ 1 \\\end{bmatrix} = \begin{bmatrix} B & A \\ A & C \\\end{bmatrix}\begin{bmatrix} \frac{\delta_1}{2} \\ \frac{\delta_2}{2} \\\end{bmatrix}$$

$$ we \space\space denote \space\space A = 1_NΣ^{-1}\mu, \space\space B = \mu^TΣ^{-1}\mu, \space\space C = 1_N^TΣ^{-1}1_N $$

위 두식을 연립하면 A,B,C에 대한 관계식이 도출되고 다음과 같다.
$$ w^* = Σ^{-1} (\frac{-C\mu_0+A}{A^2 - BC}\mu + \frac{A\mu_0 - B}{A^2 - BC}1_N ) $$

최종 목표값인 포트폴리오 최적화 가중치를 $w^*$라고 하자. 이를 구하기 위해 라그랑주 함수를 $w$에 대해 편미분하고 이 값이 0과 같다고 놓으면 다음과 같다.

$$ \frac {\partial L(w^*, \delta)}{\partial w} = 2Σw^* - \delta(\mu - \mu_f1_n) = 0 $$

$$ ⇒ {\sigma^*}^2  = {w^*}^TΣw^* = \frac {{\mu^{**}}^2}{{\mu^*}^TΣ\mu^*}$$

$$ \therefore w_p = \lambda_p^{-1}Σ^{-1}\mu^* = \lambda_p^{-1}B $$

(3)에서 구한 최적 포트폴리오의 표준편차를 위 식에 대입하면 다음과 같다.

$$ Sharpe_p = \frac {\mu^{**}}{\sigma_p} = \sqrt {{\mu^*}^TΣ\mu^*} $$

최적 포트폴리오 p의 샤프비율이 위의 식을 통해 일정한 값을 가짐을 보였다. 따라서 모든 최적 포트폴리오의 수익률 - 위험 비율, 즉 샤프 비율은 가중치 $w$가 무슨 값이든 일정하다.

결과적으로 Tangency Portfolio의 CAL은 다음과 같다.
$$ \mu_p = Sharpe_T\sigma_p + r_f $$

### 3.

1. 임의의 $\alpha$에 대해 자산 $i$의 투자 비중은 $\alpha$, 시장 포트폴리오 $M$의 투자 비중은 $(1-\alpha)$인 포트폴리오를 생각하면, $\alpha$ 값이 변함에 따라 포트폴리오는 기대수익률-표준편차 평면 상에 곡선으로 그려진다. 이 가능해 집합 위의 $\alpha = 0$인 점이 시장 포트폴리오 $M$에 해당한다. 
2. 이때 무위험 이자율과 시장 포트폴리오 $M$으로 결정되는 직선을 평면상에 그리면 이는 자본시장선(CML)이 되는데, CML 위에 있는 포트폴리오는 가능해 집합의 효율적인 경계선이 된다는 CML의 정의에 따라 이 직선은 앞서 그린 가능해 집합과 점$M$에서 접한다.
3. 종합하면 CML과 임의의 $\alpha$에 따라 생성된 가능해 집합과의 접선 포트폴리오는 시장 포트폴리오 M이고, 이 때의 접선은 자연스럽게 CML이 된다.

### 4.

#### (1)
임의의 $\alpha$에 대해 자산 $i$의 투자 비중은 $\alpha$, 시장 포트폴리오 $M$의 투자 비중은 $(1-\alpha)$인 포트폴리오를 생각하자. 포트폴리오의 기대수익률과 표준편차는 다음과 같다.
$$ \bar r_\alpha = \alpha\bar r_i + (1-\alpha)\bar r_M$$
$$ \sigma_\alpha = [\alpha^2\sigma^2_i + 2\alpha(1-\alpha)\sigma_{iM} + (1-\alpha)^2\sigma^2_M]^{1/2}$$

3번에서의 설명처럼 $\bar r - \sigma$ 평면 상에서 그려진 곡선과 CML은 $\alpha$가 0을 지날 때, 점 $M$에서 접하므로 점 $M$에서 CML과 곡선의 기울기는 같다.
이를 이용하여 앞선 식을 $\alpha$에 대해 미분하면 다음과 같다.
$$ \frac {d\bar r_\alpha}{d\alpha} = \bar r_i - \bar r_M $$

$$ \frac {d\sigma_\alpha}{d\alpha} = \frac {\alpha\sigma_i^2 + (1-2\alpha)\sigma_{iM} + (\alpha - 1)\sigma_M^2}{\sigma_\alpha} $$

$\alpha = 0$을 대입하면 다음과 같다.
$$ \frac {d\sigma_\alpha}{d\alpha} = \frac {\sigma_{iM} - \sigma_M^2}{\sigma_M} $$

$$ ⇒ \frac {d\bar r_\alpha}{d\sigma_\alpha} = \frac {(\bar r_i - \bar r_M)\sigma_M}{\sigma_{iM} - \sigma_M^2} $$

점 $M$에서 곡선의 기울기는 CML의 기울기와 같다는 것을 식으로 표현하면 다음과 같다.

$$ \frac {(\bar r_i - \bar r_M)\sigma_M}{\sigma_{iM} - \sigma_M^2} = \frac {\bar r_M - r_f}{\sigma_M} $$

마지막으로, 위 식을 $\bar r_i$에 대해 정리하면 다음과 같다.
$$ \bar r_i = r_f + (\frac {\bar r_M - r_f}{\sigma_M^2})\sigma_{iM} = r_f + \beta_i(\bar r_M - r_f) $$

이는 CAPM 공식과 일치한다.
