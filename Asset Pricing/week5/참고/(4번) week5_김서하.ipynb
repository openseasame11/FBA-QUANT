There is a portfolio consisting of one risky asset (75%) and a T-bill fund (25%).

--------------------------------

Risk-free rate = 5%.

Market expected return = 13%.

Beta of the risky asset = 2.

--------------------------------

Using the information above, find the future payoff of the portfolio, assuming to have invested $100 in the first period. Also, find the expected return of the risky asset.

A.

We can compute the beta of the portfolio, 2 * 0.75 = 1.5.



Using the above pricing formula, we can compute 100 = X / (1.05 + 1.5 * (0.13-0.05)). Therefore, X= $117.

For the expected return of the risky asset, 0.05 + 2*(0.13-0.05) = 0.21. Therefore 21%.
