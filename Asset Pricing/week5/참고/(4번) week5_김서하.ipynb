## Coursera - week1 Style & Factors

Asset management는 크게 Passive(index benchmark) / Active(out-perform, inconsistent)으로 나뉘는데, factor investing은 그 둘의 특징을 모두 가지고 있는 방식이다.
Factor은 개별 자산 밖에서 자산의 수익률에 영향(risk premium)을 주는 요인을 말한다. Macro/ statistical factor일 수 있고, 또는 intrinsic/style factor일 수도 있다.

가장 기본적인 모형으로 CAPM을 볼 수 있는데, E(rm)-rf 즉 시장수익률을 하나의 factor로 보고, β를 factor coefficient로 보는 single factor model이다. 그러나 실제로는 value/growth, momentum 등의 수많은 anomaly들이 존재한다.

Multifactor model로는 Fama-French model이 있는데, 이는 market factor에 size와 value (book/price) factor까지 고려하는 3요인 모형이다.

스타일분석은 포트폴리오의 성과를 다양한 스타일의 벤치마크 포트폴리오에 비교하여 평가하는 방법이다. 



Cap weighted index를 벤치마크로 사용하는 것이 보편적인데, 위험이 조정된 성과를 내지 못한다(위험이 잘 분산된 포트폴리오가 아님)는 점에서 비효율적이라는 한계가 존재한다. 이를 해결할 방법은 위험을 효과적으로 분산시키는 smart weighted benchmark 인데, equallty-weighted benchmarks, minimum variance benchmarks, risk parity benchmarks 등이 있다. 또한, 더 효율적으로 위험에 대해 보상받을 수 있는 방법대로 투자하지 않아 비효율적(ineffeicient factor exposures)이라는 한계도 존재한다. 이 두 문제를 해결하기 위해서는 위험에 대해 효율적으로 보상받을 수 있는, 잘 분산된 포트폴리오인 smart factor benchmarks를 구성해야한다.
