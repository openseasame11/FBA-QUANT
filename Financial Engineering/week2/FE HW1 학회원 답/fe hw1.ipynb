```python
#5.1
import sympy as sp

import time

start = time.time()

x = sp.symbols('x')

f_x = x**2+x - 1
fp_x = sp.diff(f_x)

x_init = 5
Tolerance = 0.000001

x_tmp = x_init
i = 0

while True:
    i = i + 1

    x_numeric = float(x_tmp) - float(f_x.subs(x,x_tmp))/float(fp_x.subs(x,x_tmp))

    NR_error = f_x.subs(x, x_numeric)

    if NR_error <= Tolerance:
        break
    x_tmp = x_numeric

end = time.time()
print(end - start , "s")
print(': '+ str(x_numeric))
```
