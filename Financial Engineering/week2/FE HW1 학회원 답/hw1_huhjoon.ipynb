```python
#p2
from math import factorial, exp

# Probability density of the Poisson distribution
def pois_dist(n, lamb):
    pd = (lamb ** n) * exp(-lamb) / factorial(n)
    return pd

#sol?
```

```python
#p5

import numpy as np
from timeit import default_timer as timer

 


#arr[0] is negative. 

def irr_iter(arr):
    start = timer()
    lambda0=1
    f = np.sum(arr)  
    if abs(f)<0.000001:
        return lambda0
    else:
        while abs(f)>=0.000001:
            fprime=0
            for i in range(1,len(arr)):
                fprime+=i*arr[i]*lambda0**(i-1)
            lambda0=lambda0-f/fprime
            f=0
            for i in range(len(arr)):
                f+=arr[i]*lambda0**(i)
                
    end = timer()
    print(end - start)
    return lambda0
            
    
def irr_bisection(arr, lambda0=1,left=0, right = 1000):
    
    f=0
    for i in range(len(arr)):
        f+=arr[i]*lambda0**(i) 
    if abs(f)<0.000001:
        return lambda0
    else:
        if f>0:
            return irr_bisection(arr,0.5*(lambda0+left),left,lambda0)
        else:
            return irr_bisection(arr,0.5*(lambda0+right),lambda0,right)

print("iter")    
print(irr_iter([-1,1,1]))
from timeit import default_timer as timer

print("bisection")
start=timer()
print(irr_bisection([-1,1,1]))
end=timer()
print(end-start)
```
