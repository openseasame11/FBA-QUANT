<h4> Problem 2. If you buy a lottery ticket in 50 lotteries, in each of which your chances of winning a prize of 1/100, what is the probability that you will win a prize:</h4>


Solution) Let the number of lottery wins be a random variable, $X$, then $X$ follows binomial distribution with $n=50$, and $p=\frac{1}{100}$, i.e. $X \sim B(50, \frac{1}{100})$. On the other hand, $X$ can be explained with Poisson distribution with the parameter $50\cdot \frac{1}{100} = \frac{1}{2}$.

#### (a) $P(X \leq 1)$

Binomial distribution)

$$
P(X \leq 1) = _{50}C_0 \left(\frac{1}{100}\right)^0 \left(\frac{99}{100}\right)^{50}+_{50}C_1 \left(\frac{1}{100}\right)^1 \left(\frac{99}{100}\right)^{49}=0.910565
$$

Poisson distribution)

$$
P(X \leq 1) = \frac{\left(\frac{1}{2}\right)^0 e^{-\left(\frac{1}{2}\right)}}{0!}+\frac{\left(\frac{1}{2}\right)^1 e^{-\left(\frac{1}{2}\right)}}{1!}=0.909796
$$

#### (b) $P(X=2)$

Binomial distribution)

$$
P(X=2) = _{50}C_2 \left(\frac{1}{100}\right)^2 \left(\frac{99}{100}\right)^{48}=0.046450
$$

Poisson distribution)

$$
P(X=2) = \frac{\left(\frac{1}{2}\right)^2 e^{-\left(\frac{1}{2}\right)}}{2!}=0.075816
$$

#### (c) $P(X \leq 2)$

Binomial distribution)

$$
P(X\leq 2) = P(X\leq 1) + P(X=2) = 0.910565 + 0.046450 = 0.957015
$$

Poisson distribution)

$$
P(X\leq 2) = P(X\leq 1) + P(X=2) = 0.909796 + 0.075816 = 0.985612

Solution) Note that the moment generating function for a random variable which follows a normal distribution with mean and variance of $\mu$ and $\sigma^2$ is $\exp{\left(t\mu + \frac{1}{2}\sigma^2 t^2\right)}$.

$$

#### Problem 4. Many people believe that the daily change of price of a company’s stock on the stock market is a random variable with mean 0 and variance $\sigma^2$. That is, if $S_t$ represents the price of the stock on the $t$-th day, then

$$
S_t = S_{t-1} + W_t, t \geq 1
$$

#### where $W_1, W_2, \cdots, $ are independent and identically distributed random variables with mean 0 and variance $\sigma^2$. Suppose that the stock's price today is 100 and $\sigma^2=1$.

#### (a) What can you say about the probability that the stock’s price will exceed 105 after 10 days? (Hint: Chebyshev inequality)

Solution) Note that

$$
\begin{align*}
S_1 &= S_0 + W_1
S_2 &= S_1 + W_2
&\vdots
S_{10} = S_9 + W_{10}
\end{align*}
$$
Thus,

$$
S_{10} = S_0 + W_1 + \cdots W_{10}
$$

Note that

$$
E(S_{10}) = E(S_0 + W_1 + \cdots W_{10}) = 100
$$
and
$$
Var(S_{10})= Var(S_0 + W_1 + \cdots W_{10}) = Var(W_1) + \cdots Var(W_{10}) = 10
$$

Usign Chebyshev's inequality,

$$
P(S_{10} - 100 < 5) \leq P(|S_{10} - 100| > 5 = k \sqrt{10}) \leq \frac{1}{\left(\frac{5}{\sqrt{10}}\right)^2} = \frac{2}{5}
$$

$$
\therefore P(S_{10} < 105) \leq \frac{2}{5}
$$

Here is an iterative technique that generates a sequence $\lambda_0, \lambda_1, \cdots, \lambda_k, \cdots$ of estimates that converges to the root $\bar{\lambda}>0$, solving $f(\bar(\lambda))=0$. Start with any $\lambda_0>0$ close to the solution. Assuming $\lambda_k$ has been calculated, evaluate

$$
f'(\lambda) = a_1 + \cdots + n a_n \lambda^{n-1}
$$
and define
$$
\lambda_{k+1} = \lambda_{k} - \frac{f(\lambda_k)}{f'(\lambda_k)}
$$
Try the procedure on $f(\lambda) = -1 + \lambda+\lambda^2$ starting with $\lambda_0=1$.

#### (a) Write a Python function to calculate $\bar{\lambda}$ accurate up to 0.000001 and compute the computation time.

```python
old_lambda = 1
new_lambda = old_lambda

def f(x):
    value = -1.0 + x + pow(x,2)
    
    return value

def f_prime(x):
    value = 1 + 2*x
    
    return value

start = time.time()
while (abs(f(new_lambda)) > 0.000001):
    new_lambda = old_lambda - (f(old_lambda))/(f_prime(old_lambda))
    print(new_lambda)
    old_lambda = new_lambda
print("time :", time.time() - start)
```

#### Problem 6. Given $N$ points drawn randomly on the circumference of a circle, what is the probability that they are all within a semicircle?

Solution)

At the opposite side of A, mark another point, $B$, on circumference, such that the length of $\overset{\large\frown}{AB}$ equals the half of circumference. The remaining points can lie on either $\overset{\large\frown}{AB}$ or $\overset{\large\frown}{BA}$. The probability of a point to reside in either semicircle is $\frac{1}{2}$.

Thus, the intereste probability is

$$
N\left( \frac{1}{2} \right)^{(N-1)}
