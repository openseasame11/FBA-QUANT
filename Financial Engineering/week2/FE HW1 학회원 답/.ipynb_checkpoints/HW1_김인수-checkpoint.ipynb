### Problem 4.

```python
#(C)
import matplotlib.pyplot as plt
import numpy as np

exceed = 0
for i in range(10000):
    trajectory = [100]
    for j in range(10):
        random = np.random.normal(loc=0, scale=1)
        trajectory.append(trajectory[-1] + random)
            
    if i<=10:
        plt.plot(trajectory)
        
    if trajectory[-1] >= 105:
        exceed += 1
        
plt.show()
print("10일 후 105를 넘을 확률 : {}%".format(exceed/100))
```
