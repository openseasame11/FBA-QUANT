### (c)

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(77)
```

```python
np.random.normal(0,1)
```

```python
def generate_price(M, t, s0):
  price = []
  for i in range(M):
    temp = [s0]
    for j in range(t):
      rv = temp[-1] + np.random.normal(mu, sigma)
      temp.append(rv)
    price.append(temp)
  return price
```

```python
M = 100000
t = 10
s0 = 100
mu, sigma = 0, 1
stock_price = generate_price(M, t, s0)
s10 = np.array([x[10] for x in stock_price])
```

```python
print("Probability of exceeding 105: ", round(sum(s10>105) / len(s10), 4))
```

```python
plt.figure(figsize=(8,8))
plt.plot(stock_price[:10])
plt.title("Stock Price", fontsize=18)
plt.xlabel("Time", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.show()
```

### (a)

```python
from time import time
import numpy as np
```

```python
# 원하는 함수 
def target_function(x):
  return (x**2+x-1)

# 미분한 함수
def de(x):
  return (2*x+1)

def find_sol(init_val, tol, target_function, de):
  start_time = time()
  while abs(target_function(init_val)) > tol:
    init_val = init_val - target_function(init_val) / de(init_val)
  return init_val
```

```python
init_val = 1
tol = 10**(-6)

start_time = time()
val = find_sol(init_val, tol ,target_function, de)
print("Soluction: ", val)
print("Time Taken in seconds: ", time() - start_time)
```

### (b)

```python
def bisection(f, a, b, tol): 
    m = (a + b)/2
    if np.abs(f(m)) < tol:
        return m
    elif np.sign(f(a)) == np.sign(f(m)):
        return bisection(f, m, b, tol)
    elif np.sign(f(b)) == np.sign(f(m)):
        return bisection(f, a, m, tol)
```

```python
a, b = 0, 1
tol = 10 ** (-6)

start_time = time()
val = bisection(target_function, a, b, tol)
print("Soluction: ", val)
print("Time Taken in seconds: ", time() - start_time)
```
