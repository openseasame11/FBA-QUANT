### 1 - (a)

```python
# 수치 대입 함수
def find_lambda_bisection(start, end, d):
    while True:
        x = (start + end) / 2
        y = (d['F'] / (1 + (x/d['m']))**d['n']) + sum([(d['C']/d['m']) / ((1+(x/d['m']))**n) for n in range(d['n'])])
        if (y > d['P'] - 0.0000001) and (y < d['P'] + 0.0000001):
            return x
        if y > d['P']:
            start = x
        elif y < d['P']:
            end = x
    return x

# 이자율 도출
condition1 = {'F': 1000, 'n': 10, 'P': 950, 'C':40, 'm':1}
r= find_lambda_bisection(0, 1, condition1)

# 현재가치
face_value = 1000
n = 10
zc_price = face_value / (1+r)**n

# 출력
print(f'The price of a zero-coupon bond : ${zc_price:.2f}')
```

```python
def rabbit_stair(n):
    if (n == 1) or (n == 2):
        return n
    elif n == 3:
        return 4
    else:
        return rabbit_stair(n - 3) + rabbit_stair(n - 2) + rabbit_stair(n - 1)

stairs = 4

print(f'계단이 {stairs}칸일 때, {rabbit_stair(stairs)}만큼의 경우의 수가 존재한다.')
```
