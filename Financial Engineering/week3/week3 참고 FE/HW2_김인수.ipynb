```python
# f(1) = 1, f(2) = 2, f(3) = 4이고 f(4)부터는 f(n-1) + f(n-2) + f(n-3)과 같다.    

def rabbit(n): # n ≥ 0
    if n == 0:
        return 1
    elif n == 1:
        return 1
    elif n == 2:
        return 2
    else:
        return rabbit(n-1) + rabbit(n-2) + rabbit(n-3)
```

```python
print(rabbit(0))
print(rabbit(1))
print(rabbit(2))
print(rabbit(3))
print(rabbit(4))
```
