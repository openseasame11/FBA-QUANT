## Problem 3.

Suppose that $u=e^w$, where $w$ is normal with expected value $\overline{w}$ and variance $\sigma^2$. Then,

$$
\overline{u}=\frac{1}{\sqrt{2\pi \sigma^2}}\int^\infty_{-\infty} e^we^{\frac{-(w-\overline{w})^2}{2\sigma^2}}dw
$$

Show that

$$
w-\frac{(w-\overline{w})^2}{2\sigma^2}=-\frac{1}{2\sigma^2}[w-(\overline{w}+\sigma^2)]^2+\overline{w}+\frac{\sigma^2}{2}
$$

Use the fact that

$$
\frac{1}{\sqrt{2\pi \sigma^2}}\int^\infty_{-\infty} e^{\frac{-(w-\overline{w})^2}{2\sigma^2}}dw=1
$$

to evaluate $\overline{u}$.

Solution) By doing arithmetic works,

$$
\begin{align*}
w-\frac{(w-\overline{w})^2}{2\sigma^2}&=-\frac{w^2-2(w+\sigma^2)\overline{w}+\overline{w}^2}{2\sigma^2}\\
&=-\frac{1}{2\sigma^2}[w-(\overline{w}+\sigma^2)]^2+\overline{w}+\frac{\sigma^2}{2}\\
\end{align*}
$$
