## Asset Diversification
위험은 두 가지 종류로 분류될 수 있는데, 제거할 수 없는 market(systematic) risk와, 분산 투자로 제거가능한 unique(firm-specific, unsystematic) risk이다.
Effective diversification이란 위험을 최소화하는 것을 의미하는데, 이는 각 증권들의 투자비중을 조절함으로써 달성할 수 있다. 투자비중은 꼭 0과 1 사이가 아닐 수 있는데, 예를 들어 채권에 1이상, 주식에 0이하를 투자한다면 주식 숏+채권매수 포지션으로 볼 수 있다.



위의 그림과 같이 상관계수가 1보다 작아지는 순간 두 자산(●)을 이은 직선이 굽어지면서 분산효과 (=포트폴리오 sd가 각 자산 sd의 가중합보다 작아지는 효과)가 생긴다. 1일 때에만 포트폴리오의 sd= 각 자산 sd의 가중합, 즉 분산효과가 존재하지않는다. 상관계수가 낮아질수록 분산효과가 크다. ρ=-1이라면 sd가 0이되는 조합을 찾을 수 있다.

상관계수가 특정 값으로 고정되어있을 때, 우리는 portfolio opportunity set 선상에서 MVP minimum variance portfolio(sd가 최소화되는 점)를 찾을 수 있다.

Asset allocation 단계로 넘어가면, risky portfolio 내의 자산 비중을 정하여 최적 포트폴리오를 구성할 수 있다. 특정 상관계수의 portfolio opportunity set을 그린후, 무위험수익률로부터 이 set(마코위츠모형에 따르면 efficient frontier, 가장 좌상향의 지점을 잇는 선)에 접하는 CAL을 구한다. (∵접선일때 샤프비율이 가장 높다) 접점이 바로 위험자산의 최적 포트폴리오가 된다. 이때 투자 비중값은 미분을 통하여, 샤프비율 공식 최댓값을 구하는 방식으로 계산할 수 있다.



위 계산을 통해 위험 자산 포트폴리오 내 투자비중을 구했다면, 이제 개인의 무차별곡선을 활용하여, 개인의 효용을 극대화하는 위험자산 P와 무위험자산 사이의 투자비중을 구할 수 있다.



최종적으로 optimal complete portfolio는 C점이 된다.

이처럼 두 가지 단계로 나누어 투자 결정을 하는 것을 Separation Property라고 한다. (1) technical한 방법으로 optimal risky portfolio를 구하고,(모든 client에게 동일한 포트폴리오) (2) 개인의 선호에 따라 capital allocation을 한다.

Diversification의 효과는 증권들간 correlation에 의해 달라진다. 상관계수가 0인경우 diversification의 효과는 무궁무진하지만, 실제로는 상관계수가 양수인 경우가 대부분이다.
상황을 간소화해서 살펴봤을 때 포트폴리오의 위험은 다음과 같이 계산된다.



ρ=0 이라면 n이 커질수록 위험이 0에 가까워지지만, ρ>0인 순간 위험은 ρσ^2 (=체계적 위험의 크기)으로 수렴한다.

<Risk Pooling, Risk Sharing, Time Diversification>

Risk pooling은 투자 자산을 늘리는 방식이다. 평균 payoff의 위험은 줄어들지만, 총 payoff의 위험은 더 커진다. 반면 Risk sharing은 정해진 예산 내에서 투자자산의 종류를 늘려 위험을 줄이는 방식이다. 특정 증권으로부터 기인하는 위험을 제한하는 효과가 있다. Time Diversification에서도 마찬가지이다. 특정 금액을 매년 동일하게 투자하는 것은 risk pooling, 총 위험을 증가시킨다. 반면 특정 금액을 투자기간만큼 쪼개어 길게 투자하는 것이 risk sharing, 총 위험을 줄이는 방식이다.

## Index Models
앞서 살펴본 마코위츠 모형은 애초에 추정해야하는 값들이 너무 많고, 추정한 상관계수의 값들이 불안정할 수 있다는 단점이 있었다. 추정 모수를 단순화하여, risk를 sysematic& firm-specific source 두 가지로 분해한 것이 single factor model이다. ri=E(ri)+βi * m+ei로 표현되는데, 여기서 m과 ei는 상관관계가 존재하지 않는다. 따라서 총분산은 systematic risk(βi^2 * σm^2)+ firm-specific risk(σei^2)으로 분해가 된다. 여기서 공통요인(m)을 market index로 사용한 것이 single index model이다.

Ri(t)=αi+βi*Rm(t)+ei. (αi: non market premium/ βi: security's sensitivity to the market index)




이 모델은 단순하고 계산이 편리하다는 장점이 있지만, 그만큼 불확실성 요인을 oversimplify한다(ex. 산업별 위험요인은 포착 불가)는 단점이 있다. 또한, security 간 residual correlation이 있는 경우를 마코위츠 모형과는 달리 고려하지 못한다는 단점이 있다. 이를 해결하기 위해서는 multi-index model이 사용되기도 한다.


Single Index Model을 개별 증권이 아닌 포트폴리오로 생각해보면, 총 위험은 다음과 같이 표현된다. 



포트폴리오를 구성하는 개별 증권들의 비중이 모두 같다고 가정하면, systematic risk는 포트폴리오의 베타(개별 증권의 베타 가중합(=평균))에 의해 값이 결정되며, unsystematic risk는

 에 의해 n이 커질수록 0에 가까워진다, 즉, 비체계적 위험은 분산되어 사라진다.

Index model regression은 SCL security characteristic line 증권특성선으로 표현할 수 있다. α, β의 값을 추정하여 통계적으로 검정한다. (R square, p value 등)

>Portfolio Construction

포트폴리오 구성을 위한 방법에는 트레이너-블랙 모형이 있다. 베타를 추정하는 방법은 대체로 표준화 되어있기에, 각 증권의 투자 가치를 결정 짓는 것은 알파의 값이 된다. 양의 알파값을 가진 증권은 저평가되어있다는 것을 뜻하며, 따라서 포트폴리오 내 비중이 늘어야한다.

정보비율이란 비체계적위험에 대한 알파(초과수익률)의 비율이다. 정보비율이 0이 아닌 자산에 투자하면(positive를 롱하든 negative를 숏하든), 포트폴리오의 알파를 증가시킨다는 점에서 긍정적인 효과가 있다. 포트폴리오의 residual variance를 증가시키는 경우에는 부정적인 효과를 가져올 수 있다.    
    
>Optimal Risky Portfolio
    
리스크 프리미엄과 공분산 matrix를 이용하면 portfolio maximization을 유도할 수 있다. 앞서 살펴본 equally-weighted portfolio가 아닐 때, Sharpe Ratio를 극대화하는 각 증권의 가중치 값을 결정하는 것이 목표이다. 결론적으로 optimal risky portfolio는 active portfolio와 market index (passive) portfolio 두 가지로 구성이 된다.

## Coursera; Analyzing Returns

증권의 수익률을 분석할 때 평균수익률만 보는 것은 잘못된 분석결과를 낼 수 있기에, volatility 분산을 같이 보아야 한다. (여기서 수익률은 가격의 변화분 뿐만 아니라 배당금을 포함하여 생각해야한다.) 분산은 demeaned return들을 제곱해서 모두 더하는 식으로 구한다. comparable한 값을 구하기 위해서는 분산을 루트 씌운 값인 표준편차를 사용한다. 또한 기간이 다르면 바로 비교할 수 없으므로, sqrt(p)값을 곱해서 annualize한 뒤 비교한다. Risk adjusted measures를 구할 때에는 초과수익률(return - riskfreerate)을 표준편차로 나눈다. 이것이 샤프비율이다.

Maximum DrawDown은 최악의 상황에서 잃을 수 있는 최대 금액 (worst possible return)을 의미한다. drawdown은 previous highest value (peak)와 wealth index 사이의 거리이다. drawdown의 time-series를 plot하면, max drawdown을 쉽게 볼 수 있다. Calmar Ratio는 annualized return을 지난 36개월간 maximum drawdown으로 나눈 값이다.


그러나 사실 time series returns는 정확히 정규분포를 따르지 않는다. 실제 데이터를 보면 gaussian 가정보다 extreme한 변동이 자주 일어난다. 따라서 평균(1)과 분산(2)을 넘어서 higher-order moments(skewness(3)&kurtosis(4))를 살펴보아야한다. Gaussian 가정(skewness=0, kurtosis=3)과는 달리, 실제 데이터는 skewness가 음수의 값을 가지며 kurtosis는 3보다 큰 값 (fat tail)을 가진다. -> 정확히 말하자면 실제데이터가 negatively-skewed일 때 분산으로 위험을 측정하면 tail risk를 과소평가해버릴 수 있다

Volatility를 넘어서, 더 extreme한 deviation을 살펴보기 위해서는 semi-deviation라는 개념을 사용한다. 평균보다 낮은 수익률의 경우를 살펴보고, 그 수익률들의 변동성을 살펴보기에, Gain보다 Loss에 중점을 둔다는 점에서 의미가 있다. VaR (value-at-risk)은 정해진 기간동안 발생할 수 있는 가장 큰 손실금액을 의미한다. 특정 신뢰수준 하에서 계산된다.

VaR을 넘어서는 손실의 값을 구하기 위해서는 conditional VaR (VaR을 넘어섰을 때 예상되는 손실 금액)을 구한다.



VaR을 계산하는 방법에는 네 가지가 있는데, 상황에 따라 맞는 걸 선택해서 사용한다

(1) Historical methodology: 과거 변동 분포를 사용한다. 샘플 period에 크게 좌지우지된다. (2) Parametric Gaussian: 가우스(정규)분포를 가정해서 계산한다. Fat tail로 인해 리스크를 과소평가할 수 있다. (3) Parametric Non Gaussian VaR: 여러 종류. (4) Cornish-Fisher VaR(Semi-Parametric Approach): 가우스 분포와 non-가우스 분포의 alpha quantile을 연결한다. 샘플risk와 모델risk를 적절히 조율해준다는 장점이 있다.

A.

Systematic risk of P = 1.2^2 * 0.02^2 = 0.000567
Total risk of P = 0.04^2 = 0.0016
THerefore,

```python
## Drawdowns
import pandas as pd

me_m = pd.read_csv("Portfolios_Formed_on_ME_monthly_EW.csv", header=0, index_col=0, parse_dates=True, na_values=-99.99)
rets = me_m[['Lo 10','Hi 10']]
rets.columns = ['SmallCap','LargeCap']
rets = rets/100
rets.plot.line()
```

```python
rets.index
rets.index= pd.to_datetime(rets.index, format="%Y%m")
rets.index = rets.index.to_period('M') #series.method()
rets.head()
```

```python
# computing drawdowns
# 1. compute wealth index
# 2. compute previous peaks
# 3. compute drawdown - which is the wealth value as a percentage of the previous peak
wealth_index = 1000*(1+rets["LargeCap"]).cumprod()
wealth_index.head()
```

```python
wealth_index.plot()
```

```python
previous_peaks = wealth_index.cummax()
previous_peaks.plot()
```

```python
drawdown = (wealth_index - previous_peaks)/previous_peaks
drawdown.plot()
```

```python
drawdown.head()
```

```python
drawdown.min()
```

```python
drawdown["1975":].idxmin() #minimum drawdown값을 갖는 index
```

```python
def drawdown(return_series: pd.Series):
    """
    Takes a time series of asset returns
    Computes and returns a DataFrame that contains:
    the wealth index
    the previous peaks
    percent drawdowns
    """
    wealth_index = 1000*(1+return_series).cumprod()
    previous_peaks = wealth_index.cummax()
    drawdowns = (wealth_index - previous_peaks)/previous_peaks
    return pd.DataFrame({
        "Wealth": wealth_index,
        "Peaks": previous_peaks,
        "Drawdown": drawdowns
    })
```

```python
def var_historic(r, level=5):
    """
    VaR Historic
    """
    if isinstance(r, pd.DataFrame):
        return r.aggregate(var_historic, level=level)
    elif isinstance(r, pd.Series):
        return -np.percentile(r, level)
    else:
        raise TypeError("Expected r to be Series or DataFrame")
```

```python
## a.

stocks= ['AAPL','MSFT','AMZN','GOOGL','NVDA']

df_stocks=yf.download(stocks,
                   start='2018-01-01',
                   end='2023-01-01',
                   progress=False)

df_stocks=df_stocks['Adj Close'].pct_change().dropna()
daily_returns=pd.DataFrame(df_stocks)
daily_returns
```

```python
##b.

annual_ret = []

for i in stocks:
    annual_ret.append((((1+daily_returns[i]).prod())**(1/1258)-1)*252)
annual_ret
```

```python
daily_cov = daily_returns.cov()
annual_cov = daily_cov * 252
annual_cov
```

```python
## c.

port_ret = []
port_risk = []
port_weight = []

for _ in range(20000):
    weights = np.random.random(len(stocks))
    weights /= np.sum(weights)
    temp_returns = np.dot(weights, annual_ret)
    temp_risk = np.sqrt(np.dot(weights.T, np.dot(annual_cov, weights)))
    port_ret.append(temp_returns)
    port_risk.append(temp_risk)
    port_weight.append(weights)

plt.scatter(port_risk, port_ret, marker=".", c="purple")
plt.xlim([0, 0.52])
plt.ylim([0, 0.35])
plt.show()
```

```python
portfolio = []
for i in range(len(port_risk)):
    portfolio.append([port_risk[i],port_ret[i]])

optimal_port_S = 0
optimal_port = []

for p in portfolio:
    Sharpe = (p[1]-0.02)/p[0]
    if Sharpe > optimal_port_S:
        optimal_port_S=Sharpe
        optimal_port=[]
        optimal_port.append([p[0],p[1]])
        
optimal_port=optimal_port[0]

print(optimal_port)
print(optimal_port_S)
```
