Q.

Assume there exists a portfolio, investing 75% in stock A and 25% in stock B. Beta of A is 1.4, and beta of B is 0.6. The standard deviation is 4% for this portfolio, and 2% for the market portfolio. What is the ratio of the portfolio's systematic risk over the total risk?

A.

Portfolio's beta = 1.4*0.75 + 0.6*0.25 = 1.2
Systematic risk of P = 1.2^2 * 0.02^2 = 0.000567
Total risk of P = 0.04^2 = 0.0016
THerefore,
systematic risk/ total risk = 0.000567/0,0016 = 35.44%

```python
def drawdown(return_series: pd.Series):
    """
    Takes a time series of asset returns
    Computes and returns a DataFrame that contains:
    the wealth index
    the previous peaks
    percent drawdowns
    """
    wealth_index = 1000*(1+return_series).cumprod()
    previous_peaks = wealth_index.cummax()
    drawdowns = (wealth_index - previous_peaks)/previous_peaks
    return pd.DataFrame({
        "Wealth": wealth_index,
        "Peaks": previous_peaks,
        "Drawdown": drawdowns
    })
```

```python
def var_historic(r, level=5):
    """
    VaR Historic
    """
    if isinstance(r, pd.DataFrame):
        return r.aggregate(var_historic, level=level)
    elif isinstance(r, pd.Series):
        return -np.percentile(r, level)
    else:
        raise TypeError("Expected r to be Series or DataFrame")
```

```python
## a.

stocks= ['AAPL','MSFT','AMZN','GOOGL','NVDA']

df_stocks=yf.download(stocks,
                   start='2018-01-01',
                   end='2023-01-01',
                   progress=False)

df_stocks=df_stocks['Adj Close'].pct_change().dropna()
daily_returns=pd.DataFrame(df_stocks)
daily_returns
```

```python
##b.

annual_ret = []

for i in stocks:
    annual_ret.append((((1+daily_returns[i]).prod())**(1/1258)-1)*252)
annual_ret
```

```python
daily_cov = daily_returns.cov()
annual_cov = daily_cov * 252
annual_cov
```

```python
## c.

port_ret = []
port_risk = []
port_weight = []

for _ in range(20000):
    weights = np.random.random(len(stocks))
    weights /= np.sum(weights)
    temp_returns = np.dot(weights, annual_ret)
    temp_risk = np.sqrt(np.dot(weights.T, np.dot(annual_cov, weights)))
    port_ret.append(temp_returns)
    port_risk.append(temp_risk)
    port_weight.append(weights)

plt.scatter(port_risk, port_ret, marker=".", c="purple")
plt.xlim([0, 0.52])
plt.ylim([0, 0.35])
plt.show()
```

```python
portfolio = []
for i in range(len(port_risk)):
    portfolio.append([port_risk[i],port_ret[i]])

optimal_port_S = 0
optimal_port = []

for p in portfolio:
    Sharpe = (p[1]-0.02)/p[0]
    if Sharpe > optimal_port_S:
        optimal_port_S=Sharpe
        optimal_port=[]
        optimal_port.append([p[0],p[1]])
        
optimal_port=optimal_port[0]

print(optimal_port)
print(optimal_port_S)
```
