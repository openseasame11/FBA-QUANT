**CH6**  
수익률(Rate of Return) = $ \frac{X_1-X_0}{X_0}$  
Short(숏포지션)은 Long(롱포지션)과 수익의 방향이 반대가 된다  
포트폴리오의 수익률은 다음과 같이 표시할 수 있다
$ r = \sum_{i=1}^{n}{w_i}{r_i} $
$ (\sum_{i=1}^{n}{w_i} = 1) $  
즉, 포트폴리오의 수익률은 각 자산의 비중을 가중평균한 것이라고 할 수 있다  
$ \\ $  
현재 시점에서 미래의 자산 가격(수익률)을 정확하게 알 수는 없다. 미래의 자산 가격(수익률)을 확률 변수로 생각하고, 다음과 같은 것들을 추정할 수 있다.  
기대수익률  
변동성과 상관계수(공분산 행렬)  
$ \\ $   
각 자산의 기대수익률과 공분산행렬을 추정한 결과를 바탕으로 포트폴리오의 기대수익률과 공분산 행렬을 이용해 포트폴리오의 기대수익률과 분산을 추정하면 다음과 같다.  
$ E[r] = \sum_{i=1}^{n}{w_i}{E[r_i]} $  
$\sigma ^ {2} = \sum_{i,j=1}^{n}{w_i}{w_j}{\sigma_{ij}} $  
포트폴리오의 분산을 구하는 식에서도 알 수 있듯이 자산을 분산투자함으로써 비체계적 위험을 낮춰 포트폴리오의 변동성을 낮출 수 있다  
$ \\ $   
이제 자산이 두 개인 경우를 기준으로 생각해보자.  
자산이 두개이고, 자산의 비중을 각각 ${w_1}, 1-{w_1}$라고 하자  
포트폴리오의 기대수익률= ${w_1}{\bar{r_1}}+ (1-{w_1}){\bar{r_2}} $  
포트폴리오의 변동성 = $ \sqrt{{w_1}^2{\sigma_1}^2 + (1-{w_1})^2{\sigma_2}^2 + 2 * {p} * {w_1}(1-{w_1}){\sigma_1}{\sigma_2}} $  
상관계수가 1일 때, 가장 높은 변동성을, 상관계수가 -1일 때, 가장 낮은 변동성을 지닌다. 그리고 위의 식들을 통해 포트폴리오의 기대수익률과 변동성 사이의 관계를 나타낼 수 있다.   
$ \\ $   
$ \sum_{i=1}^{n}{w_i} = 1 $ 이라는 제약 조건 하에서 구축가능한 모든 포트폴리오들을 feasible set, 구축가능한 모든 포트폴리오들 중 각각의 기대수익률에 대해 가장 낮은 변동성을 지니는 포트폴리오들을 효율적투자곡선(efficient frontier)라고 한다.  
$ \\ $  
주어진 기대수익률 하에서 변동성이 최소가 되는 포트폴리오를 라그랑주 승수를 이용해 풀 수 있다.  (수식은 생략)  
$ \\ $  
Two Fund Theorm  
효율적투자곡선 위의 포트폴리오들은 효율적 투자곡선 위에 있는 다른 두 포트폴리오들의 조합으로 이뤄진다.  
One Fund Therom  
만약, 무위험자산을 고려한다면 효율적투자곡선은 완전히 달라지게 된다. 효율적투자곡선은 곡선의 형태가 아니라 샤프비율을 최대로 하는 포트폴리오와 무위험자산만으로 이뤄진 포트폴리오들의 조합으로 이뤄진다.


**Week 2**  
기대수익률과 변동성 하에서 자산 비중을 다르게 함으로써 다양한 포트폴리오 생성 가능  
이 때, 정해진 기대수익률 하에서 가장 낮은 변동성을 지니는 포트폴리오의 모음을 효율적 투자곡선이라고 한다.  
$ \\ $  
만약, 무위험 자산을 포함시킨다면 효율적투자곡선은 샤프비율을 최대로 하는 포트폴리오와 무위험자산만으로 이뤄진 포트폴리오들의 조합으로 이뤄지게 된다. 따라서 직선의 형태로 나타나게 된다.  
$ \\ $  
MVO는 이론적으로 견고하지만 사용되는 값이 조금만 변해도 결과가 크게 바뀔 수 있다는 한계점이 존재한다. 따라서 GMV(글로벌최소분산)이나 EW(동일가중) 포트폴리오를 사용할 수 있다.




**Week 3**  
분산투자는 비체계적 위험을 제거할 수는 있지만 체계적 위험을 제거할 수는 없다. 따라서 Downside Protection을 하고 싶다면 헤지를 해야한다.  CPPI가 대표적인 방법이다.  
$ \\ $  
CPPI(고정 비율 포트폴리오 보험 전략)은 포트폴리오의 가치가 미리 정한 하한보다 내력가지 않도록 설계된 모델이다. CPPI 전략은 포트폴리오의 가치(P) 와 하한(B)의 차이인 쿠션(C)와 레벌리지 승수(M)에 의해 정의된다.   
$ \\ $  
위험자산에 대한 가중치는 M*C를 부여한다. 즉, 포트폴리오의 가치가 상승해 C가 커지면 위험자산에 높은 가중치를, 낮아지면 C가 작아져 위험 자산에 낮은 가중치를 부여하게 된다. 만약, C가 0이 되면 무위험자산에만 투자하게 된다.  
$ \\ $   
CPPI는 시장이 빠르게 움직이는 상황에서는 제대로 작동하지 못한다는 단점이 있다. 즉, 갭 리스크가 있다. CPPI 전략은 적은 비용으로 효율적인 하방 보호를 할 수 있다는 장점이 있다. 또한 CPPI를 응용한 전략들을 구사할 수 있다.  
$ \\ $  
주식의 수익률은 랜덤워크를 따른다고 할 수 있다. 즉, 주식의 수익률은 정규분포를 따르고, 이 때, 자기상관이 0이고, 변동성이 시간에 비례하여 커진다.  
이러한 특징들을 반영해 Brownian Motion Process를 사용해 수익률을 모델링 할 수 있다.  
주식의 수익률을 모델링한 후, 몬테카를로 시뮬레이션을 사용해 과정을 여러 번 반복할 수 있다. 다양한 전략들을 시뮬레이션하고, 그 결과를 평가할 수 있다.

자산 X의 기대수익률은 9%, 변동성(표준편차)는 20%, 자산 Y의 기대수익률은 4%,변동성(표준편차)는 8% 라고 하자.  
각 자산의 비중을 $w_1, w_2$ 라고 할 때, 포트폴리오의 기대수익률을 $w_1, w_2$에 관한 식으로 표현하여라.

답: $w_1 * 0.09 + w_2 *0.04 $

###(b)  
자산 간의 상관관계가 -1일 때, 포트폴리오의 변동성(표준편차)를 $w_1, w_2$에 관한 식으로 표현하여라.

답: $ \sqrt{{w_1}^2*0.2^2 + {w_2}^2*0.08^2 + 2 * (-1) * {w_1}{w_2}*0.2*0.08} $ = $ |{w_1}*0.2 - {w_2}*0.08|$

###(c)  
자산 간의 상관관계가 1일 때, 포트폴리오의 변동성(표준편차)를 $w_1, w_2$에 관한 식으로 표현하여라.

답: $ \sqrt{{w_1}^2 * 0.2^2 + {w_2}^2 * 0.08^2 + 2 * {w_1}{w_2} * 0.2 * 0.08} $ = $ |{w_1} * 0.2 + {w_2} * 0.08|$

#### CPPI and Drawdown Constraint

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import edhec_risk_kit_119 as erk

# Load the industry returns and the total market index we previously created
ind_return = erk.get_ind_returns()
tmi_return = erk.get_total_market_index_returns()
```

```python
import pandas as pd
import numpy as np

risky_r = ind_return["2000":][["Steel", "Fin", "Beer"]]
# Assume the safe asset is paying 3% per year
safe_r = pd.DataFrame().reindex_like(risky_r)
safe_r.values[:] = 0.03/12 # fast way to set all values to a number
start = 1000 # start at $1000
floor = 0.80 # set the floor to 80 percent of the starting value
```

```python
def compound1(r):
    return np.expm1(np.log1p(r).sum())

def compound2(r):
    return (r+1).prod()-1

compound1(ind_return[["Steel", "Fin", "Beer"]])
```

```python
compound2(ind_return[["Steel", "Fin", "Beer"]])
```

```python
%timeit compound1(ind_return)
```

```python
%timeit compound2(ind_return)
```

```python
# set up the CPPI parameters
dates = risky_r.index
n_steps = len(dates)
account_value = start
floor_value = start*floor
m = 3
## set up some DataFrames for saving intermediate values
account_history = pd.DataFrame().reindex_like(risky_r)
risky_w_history = pd.DataFrame().reindex_like(risky_r)
cushion_history = pd.DataFrame().reindex_like(risky_r)
```

```python
for step in range(n_steps):
    cushion = (account_value - floor_value)/account_value
    risky_w = m*cushion
    risky_w = np.minimum(risky_w, 1)
    risky_w = np.maximum(risky_w, 0)
    safe_w = 1-risky_w
    risky_alloc = account_value*risky_w
    safe_alloc = account_value*safe_w
    # recompute the new account value at the end of this step
    account_value = risky_alloc*(1+risky_r.iloc[step]) + safe_alloc*(1+safe_r.iloc[step])
    # save the histories for analysis and plotting
    cushion_history.iloc[step] = cushion
    risky_w_history.iloc[step] = risky_w
    account_history.iloc[step] = account_value
    risky_wealth = start*(1+risky_r).cumprod()
```

```python
ind = "Beer"
ax = account_history[ind].plot()
risky_wealth[ind].plot(style="k:")
ax.axhline(y=floor_value, color='r', linestyle='--')
```

```python
ind = "Fin"
ax = account_history[ind].plot()
risky_wealth[ind].plot(style="k:")
ax.axhline(y=floor_value, color='r', linestyle='--')
```

#### Random Walk Generation

```python
%matplotlib inline
import numpy as np
import pandas as pd

def gbm0(n_years = 10, n_scenarios=1000, mu=0.07, sigma=0.15, steps_per_year=12, s_0=100.0):
    """
    Evolution of Geometric Brownian Motion trajectories, such as for Stock Prices
    :param n_years:  The number of years to generate data for
    :param n_paths: The number of scenarios/trajectories
    :param mu: Annualized Drift, e.g. Market Return
    :param sigma: Annualized Volatility
    :param steps_per_year: granularity of the simulation
    :param s_0: initial value
    :return: a numpy array of n_paths columns and n_years*steps_per_year rows
    """
    # Derive per-step Model Parameters from User Specifications
    dt = 1/steps_per_year
    n_steps = int(n_years*steps_per_year)
    xi = np.random.normal(size=(n_steps, n_scenarios))
    rets = mu*dt + sigma*np.sqrt(dt)*xi
    # convert to a DataFrame
    rets = pd.DataFrame(rets)
    # convert to prices
    prices = s_0*(rets+1).cumprod()
    return prices
```

```python
def gbm(n_years = 10, n_scenarios=1000, mu=0.07, sigma=0.15, steps_per_year=12, s_0=100.0):
    """
    Evolution of Geometric Brownian Motion trajectories, such as for Stock Prices
    :param n_years:  The number of years to generate data for
    :param n_paths: The number of scenarios/trajectories
    :param mu: Annualized Drift, e.g. Market Return
    :param sigma: Annualized Volatility
    :param steps_per_year: granularity of the simulation
    :param s_0: initial value
    :return: a numpy array of n_paths columns and n_years*steps_per_year rows
    """
    # Derive per-step Model Parameters from User Specifications
    dt = 1/steps_per_year
    n_steps = int(n_years*steps_per_year) + 1
    rets_plus_1 = np.random.normal(loc=(mu*dt+1), scale=(sigma*np.sqrt(dt)), size=(n_steps, n_scenarios))
    rets_plus_1[0] = 1
    prices = s_0*pd.DataFrame(rets_plus_1).cumprod()
    return prices
```

```python
def gbm2(n_years = 10, n_scenarios=1000, mu=0.07, sigma=0.15, steps_per_year=12, s_0=100.0):
    """
    Evolution of Geometric Brownian Motion trajectories, such as for Stock Prices
    :param n_years:  The number of years to generate data for
    :param n_paths: The number of scenarios/trajectories
    :param mu: Annualized Drift, e.g. Market Return
    :param sigma: Annualized Volatility
    :param steps_per_year: granularity of the simulation
    :param s_0: initial value
    :return: a numpy array of n_paths columns and n_years*steps_per_year rows
    """
    # Derive per-step Model Parameters from User Specifications
    dt = 1/steps_per_year
    n_steps = int(n_years*steps_per_year) + 1
    rets_plus_1 = np.random.normal(loc=(1+mu)**dt, scale=(sigma*np.sqrt(dt)), size=(n_steps, n_scenarios))
    rets_plus_1[0] = 1
    prices = s_0*pd.DataFrame(rets_plus_1).cumprod()
    return prices
```

```python
import matplotlib.pyplot as plt
import numpy as np

def show_cppi(n_scenarios=50, mu=0.07, sigma=0.15, m=3, floor=0., riskfree_rate=0.03, steps_per_year=12, y_max=100):
    """
    Plot the results of a Monte Carlo Simulation of CPPI
    """
    start = 100
    sim_rets = erk.gbm(n_scenarios=n_scenarios, mu=mu, sigma=sigma, prices=False, steps_per_year=steps_per_year)
    risky_r = pd.DataFrame(sim_rets)
    # run the "back"-test
    btr = erk.run_cppi(risky_r=pd.DataFrame(risky_r),riskfree_rate=riskfree_rate,m=m, start=start, floor=floor)
    wealth = btr["Wealth"]

    # calculate terminal wealth stats
    y_max=wealth.values.max()*y_max/100
    terminal_wealth = wealth.iloc[-1]
    
    tw_mean = terminal_wealth.mean()
    tw_median = terminal_wealth.median()
    failure_mask = np.less(terminal_wealth, start*floor)
    n_failures = failure_mask.sum()
    p_fail = n_failures/n_scenarios

    e_shortfall = np.dot(terminal_wealth-start*floor, failure_mask)/n_failures if n_failures > 0 else 0.0

    # Plot!
    fig, (wealth_ax, hist_ax) = plt.subplots(nrows=1, ncols=2, sharey=True, gridspec_kw={'width_ratios':[3,2]}, figsize=(24, 9))
    plt.subplots_adjust(wspace=0.0)
    
    wealth.plot(ax=wealth_ax, legend=False, alpha=0.3, color="indianred")
    wealth_ax.axhline(y=start, ls=":", color="black")
    wealth_ax.axhline(y=start*floor, ls="--", color="red")
    wealth_ax.set_ylim(top=y_max)
    
    terminal_wealth.plot.hist(ax=hist_ax, bins=50, ec='w', fc='indianred', orientation='horizontal')
    hist_ax.axhline(y=start, ls=":", color="black")
    hist_ax.axhline(y=tw_mean, ls=":", color="blue")
    hist_ax.axhline(y=tw_median, ls=":", color="purple")
    hist_ax.annotate(f"Mean: ${int(tw_mean)}", xy=(.7, .9),xycoords='axes fraction', fontsize=24)
    hist_ax.annotate(f"Median: ${int(tw_median)}", xy=(.7, .85),xycoords='axes fraction', fontsize=24)
    if (floor > 0.01):
        hist_ax.axhline(y=start*floor, ls="--", color="red", linewidth=3)
        hist_ax.annotate(f"Violations: {n_failures} ({p_fail*100:2.2f}%)\nE(shortfall)=${e_shortfall:2.2f}", xy=(.7, .7), xycoords='axes fraction', fontsize=24)

cppi_controls = widgets.interactive(show_cppi,
                                   n_scenarios=widgets.IntSlider(min=1, max=1000, step=5, value=50), 
                                   mu=(0., +.2, .01),
                                   sigma=(0, .3, .05),
                                   floor=(0, 2, .1),
                                   m=(1, 5, .5),
                                   riskfree_rate=(0, .05, .01),
                                   steps_per_year=widgets.IntSlider(min=1, max=12, step=1, value=12,
                                                          description="Rebals/Year"),
                                   y_max=widgets.IntSlider(min=0, max=100, step=1, value=100,
                                                          description="Zoom Y Axis")
)
display(cppi_controls)
```
