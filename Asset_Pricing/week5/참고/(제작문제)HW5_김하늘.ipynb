#### 2.1 CAPM
$ \bar{r_i} - r_f = \beta_i(\bar{r_M}-r_f) $  
$ \beta_i = \frac{\sigma_{iM}}{\sigma_M^2} $  

 CAPM : single-factor(시장수익률) model 
#### 2.2 Multi-Factor Model
FAMA FRENCH Model: sort stocks into deciles based on (1) SMB (2) HML (3) MKT

- (1) SMB : "small cap stocks" - "large cap stocks"
- (2) HML : "high book/price (value)" - "low book/price (growth)"
- (3) MKT : market factor

$ E[r_i] = r_f + \beta_{i,MKT}E[r_m-r_f] + \beta_{i, SMB}E[SMB] + \beta_{i,HML}E[HML] $
