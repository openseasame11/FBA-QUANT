## Problem 4.

### $E(r_p) = 0.18$, $σ_p = 0.28$, $r_f = 0.08$

### (a) the expected value of the rate of return on his portfolio: $E(r_c)$ = 0.7$E(r_p) + 0.3r_f$ 
### the standard deviation of the rate of return on his portfolio: $σ_c$ = 0.7$σ_p$

```python
# Given values
Er_p = 0.18
σ_p = 0.28
r_f = 0.08

#Solve
Er_c = 0.7*Er_p + 0.3*r_f
σ_c = 0.7*σ_p

#Results
print(Er_c)
print(σ_c)
```

### (b) Sharpe ratio : $S$ =$ (E(r_c) - r_f) \overσ_c $
### $ (E(r_c) - r_f) \overσ_c $ = $(0.7E(r_p) + 0.3r_f - r_f) \over 0.7σ_p $ = $ (E(r_p) - r_f) \over σ_p $

```python
#Solve
S_mine = (Er_p - r_f) /σ_p
S_client = (Er_c - r_f) /σ_c

#Results
print(S_mine)
print(S_client)
```

### (c) CAL : $E(r_c) = r_f + y[E(r_p) - r_f] $ = $ r_f$ + $σ_c \over σ_p$ $ [E(r_p) - r_f] $ = $ r_f + Sσ_p $
### The slope of CAL is sharpe ratio.

```python
import numpy as np
import matplotlib.pyplot as plt

#Draw graph
σ = np.linspace(0, 5)
E_r = r_f + S_mine*σ

plt.plot(σ, E_r, label = 'CAL')
plt.xlabel('$σ$', loc ='right')
plt.ylabel('$E_(r)$', loc = 'top')
plt.xlim([0, 5])
plt.ylim([0, 2.5])

plt.legend()
plt.show()
```

### (d) $E(r_n)$ = $ 0.18y + 0.08(1-y) = 0.16$

### Since y is 1.16667, my client takes long postion in my portfolio with 80% and takes long position in the T-bill with 20%. 

```python
import sympy

#Setting
y = sympy.symbols("y")
Er_n = 0.16
f = sympy.Eq(0.18*y + 0.08*(1-y), Er_n)

#Solve
result = sympy.solve(f)

print(result)
```
