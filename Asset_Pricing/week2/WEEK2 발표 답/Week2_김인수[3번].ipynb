## 1. Summary

### 5.1 Determinants of the Level of Interest Rates
    1) 명목이자율(nominal interest rate) : 명목상의 금액이라는 관점의 이자율
       실질이자율(real interest rate) : 투자에 따른 구매력의 성장률 
    
    2) 균형 실질이자율은 수요 곡선과 공급 곡선이 만나는 교차점 E의 y값 
    
    3) 피셔방정식 : 기대 인플레이션율 E(i)와 명목, 실질 이자율과의 방정식 

### 5.2 Comparing Rates of Return for Different Holding Periods
    1) 연이자율(APR) = 기간별 수익률 * 연간 기간의 수, 실효연간이자율(EAR) : 투자자금의 실제 성장률  

    2) T의 값이 0으로 접근하면, 즉 복리기간의 수가 무수히 많아지면 CC 값이 된다.
 
### 5.4 Risk and Risk Premium
    1) 위험을 정량화하기 위해 시나리오를 작성하고 각 시나리오의 확률(가능성)에 따른 HPR을 구체화한다. 이러한 HPR과 이와 관련된 확률을 나열한 것을 확률분포라고 한다.
    
    2) 시나리오의 기대수익률 - 무위험 수익률 = 위험 프리미엄
       위험자산의 실제 수익률 - 실제 무위험 수익률 = 초과 수익률. 
    즉, 위험 프리미엄은 초과수익률의 기대값이며, 초과 수익률의 표준편차는 위험의 척도가 된다. 투자자들은 위험을 회피하는 성향이 있기 때문에 위험 프리미엄이 0이라면 투자하지 않을 것이다. 따라서 위험을 감수할만한 만족스러운 위험 프리미엄이 제공되어야 한다.  

### 5.5 Time Series Analysis of Past Rates of Return
    1) 산술평균수익률 : 하나의 관측치를 하나의 시나리오로 생각하고 각각의 확률을 1/(관측치의 개수)로 놓고 구한 기대수익률.
    
    2) 기하평균(시간가중평균)수익률 : 투자 종료기간의 누적 수익을 구하고 보유 기간에 따라 가중되어 산출된 수익률.
    
    3) 샤프비율 : 위험 프리미엄을 초과 수익률의 표준편차로 나눈 값으로 위험과 보상간의 trade-off를 수치화한 지표이다.
    

### 5.7 Deviations from Normality and Alternative Risk Measures
    1) 왜도(skew) : 분포의 비대칭성을 측정하는 척도, 3차 적률
    
    2) 첨도(kurtosis) : 확률분포의 꼬리의 뚱뚱한 정도를 측정하는 척도, 극단적 결과가 나타날 가능성을 알려줌, 4차 적률
    
    3) 앞서 설명한 표준편차 뿐만 아니라 왜도, 첨도, VaR 등 또한 위험을 측정하는 척도로 쓰이며 수익률이 정규분포를 이루는 경우 표준편차가 의미 있는 척도로 사용된다. 바꿔 말하면 수익률이 정규분포를 따르지 않으면 표준편차는 위험을 측정하기에 충분하지 않다. 따라서 중요한 것은 정규분포성을 당연한 것으로 받아들여서는 안 된다.

### 5.8 Normality and Long-Term Investments
    1) 투자 수익률의 상승 잠재력은 무한하고 하강 잠재력은 0 밑으로 떨어질 수 없기 때문에 바닥이 정해져있다. 따라서 투자 기간이 늘어날수록 양의 방향으로 꼬리가 치우쳐진 비대칭적인 패턴이 나타나고 이때 실제 분포는 로그 정규분포를 따른다.
    
    2) 다만 여러 기간 동안의 연속복리수익률은 지수함수의 곱으로 나타내지므로, 한 기간의 수익률이 정규분포를 따르면 다기간의 수익률 또한 정규분포를 따른다고 단순화 가정이 가능하다.
### 6.1 Risk and Risk Aversion
    1) 투자자의 위험회피 성향에 따라 포트폴리오의 효용수준이 달라진다. 효용점수는 기대수익률이 높을수록, 변동성이 낮을수록 높다.
    
    2) 포트폴리오의 위험이 증가하면 그만큼 기대수익률 또한 증가해야 같은 선호를 갖는 포트폴리오가 될 것이다. 이렇듯 효용가치가 같은 포트폴리오를 선으로 이으면 곡선이 그려지는데 이를 무차별 곡선이라고 한다.  
    

### 6.3 The Risk-Free Asset
    1) 무위험자산 : 실질가치 기준으로는 TIPS만이 진정한 무위험 자산이지만 일반적으로 T-bill 또한 무위험자산으로 간주한다. 실제로는 만기가 짧은 단기 금융상품들은 사실상 이자율 위험으로부터 벗어나 있고 채무불이행 위험 측면에서 거의 안전하기 때문에 무위험자산으로 활용된다.
### 6.4 Portfolios of One Risky Asset and a Risk-Free Asset
    1) 완성 포트폴리오의 위험프리미엄은 위험자산의 위험프리미엄에 포트폴리오의 위험자산 비율을 곱한 것과 같다. 완성 포트폴리오의 표준편차는 위험자산의 표준편차에 포트폴리오의 위험자산 비율을 곱한것과 같다.
    
    2) 자산배분, 즉 y값을 변화시킬 때 실현 가능한 위험-수익률 조합을 투자기회집합이라고 하고 밑 그림에서의 직선을 자본배분선(CAL)이라고 한다. CAL의 기울기(S)는 표준편차가 한 단위 증가할 때 투자가가 얻는 기대수익률의 증가분과 같다. 이 기울기는 5장에서 설명한 샤프비율과 같고 보상-변동성 비율이라고도 한다. 자본배분선 상에 있는 모든 완성 포트폴리오의 샤프비율은 동일하다.
    
    3) 만약 y가 1을 넘어, 즉 차입을 하여 위험자산에 투자한다면 자본배분선 상의 점 P 오른쪽에 위치한다. 다만 개인투자자의 경우 차입이자율이 무위험 이자율을 초과하기 때문에 자본배분선은 P를 기준으로 완만한 모습으로 꺾여 나타난다.. 

### 6.5 Risk Tolerance and Asset Allocation
    1) 동일한 자본배분선이 주어져도 개인의 위험회피 수준(효용함수)에 따라 각기 다른 위험자산 포지션을 선택한다. 위의 효용함수 식과 자본배분선 식을 연립하면 아래와 같은 식이 도출되고 이를 y에 대해 미분하여 도함수 = 0으로 놓으면 최적의 효용수준을 제공하는 위험자산의 비율(y*)을 구할 수 있다.
    
    2) 위험회피계수와 효용수준을 특정한 값들로 고정하면 표준편차가 변함에 따라 기대수익률 또한 변하는데 이를 평면상에 나타내면 다양한 무차별곡선들이 그려진다. 이 무차별곡선들과 자본배분선을 겹쳐 그릴 때 자본배분선과 접하는 무차별곡선이 존재하고 접점의 표준편차와 기대수익률이 최적의 효용수준을 만드는 위험자산비율(y*)일 때의 값과 같다. 바꿔 말하면 y*의 선택(도출)은 위험회피(무차별곡선의 기울기)와 샤프비율(투자기회집합의 기울기)에 의해 결정된다.

```python
import pandas as pd

df = pd.read_csv('C:\\Users\insu\Desktop\\6_Portfolios_2x3.csv', skiprows=15) 
df.rename(columns = {'Unnamed: 0':'Month'}, inplace=True) # 날짜 데이터의 열 이름 설정
df.loc[df['Month'] == '193001']  # 42행     많은 데이터셋 중 "monthly value weighted return"의 값만 추출하기 위해 행의 번호 찾기
df.loc[df['Month'] == '201612']  # 1085행
new_df = df.iloc[42:1086]
new_df = new_df.set_index(keys='Month')
new_df
```

```python
new_df1 = new_df.iloc[:522].astype('float') # dtype을 float으로 변환
new_df2 = new_df.iloc[522:].astype('float')

stat1 = new_df1.describe()['mean':'std'] # describe를 이용해 mean, std 값을 행으로 갖는 데이터프레임을 생성
stat1 = pd.concat([stat1,pd.DataFrame(new_df1.skew(),columns=['skew']).transpose()]) # skew , kurtosis를 새로운 행으로 추가
stat1 = pd.concat([stat1,pd.DataFrame(new_df1.kurt(),columns=['kurtosis']).transpose()])
print(stat1)

stat2 = new_df2.describe()['mean':'std']
stat2 = pd.concat([stat2,pd.DataFrame(new_df2.skew(),columns=['skew']).transpose()])
stat2 = pd.concat([stat2,pd.DataFrame(new_df2.kurt(),columns=['kurtosis']).transpose()])
print(stat2)
```

두 데이터프레임을 비교하면 수익률의 평균과 표준편차는 비교적 큰 차이를 보이지 않으나 왜도와 첨도에서 큰 차이를 보인다. 즉 위험의 정도에서 큰 차이를 보이므로 두 기간이 같은 분포를 보인다고 하기 어렵다.
