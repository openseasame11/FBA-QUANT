#### 7.1. Market Equilibrium

#### 7.2. The Capital Market Line
a. The Capital Makret Line <br/>
- Given the preceding conclusion that the single efficient fund of risky assets is the market portfolio, we can label this fund on the $\bar r - \sigma$ diagram with an $M$ for market.
- The efficient set therefore consist of a single straight line, emanating from the risk-free point and passing throught the makret portfolio. That line can be called  CML.
- Efficient assets must all lie on the line determined by the risky-free rate and the market portfolio

b. CML in mathematical terms <br/>
$$ \bar r = r_f + \frac{\bar r_M - r_f}{\sigma_M}\sigma $$ 
c. $K = \bar r_M - r_f$ is the slope of CML and this value is frequently called the price of risk<br/>
#### 7.3. The Pricing Model
a. The Capital Asset Pricing Model (CAPM)
- If the market portfolio $M$ is efficient, the expected return $\bar {r_i}$ of any asset i satifies
$$ \bar r_i - r_f = \beta_i (\bar r_M - r_f),$$
where
$$ \beta_i = \frac{\sigma_{iM}}{\sigma_M^2}$$

b. The value $\beta_i$ is referred to as the beta of an asset. When the asset is fixed in a discussion, we often just write beta without a subscript - $\beta$. An asset's beta is all that need to be known about the asset's risk characteristic to use the CAPM formula <br/>
c. $\beta_p = \Sigma_{i=1}^n w_i \beta_i$

b. Systematic Risk
- The random rate of return of asset i can be wirte as
$$ r_i = r_f + \beta_i(r_M - r_f) + \epsilon_i \;\;(A)$$
- First, thaing the expected value of (A), the CAPM says that $E(\epsilon_i)=0$. Second, taking the correlation of (A) with $r_M$, we find $cov(\epsilon_i, \sigma_M)=0$
- The variance of asset i is
$$ \sigma_i^2 = \beta_i^2\sigma_M^2 + var(\epsilon_i) $$
The first part $\beta_i^2\sigma_M^2$ is termed the systematic risk. This is the risk associated with the market whole. The second part $var(\epsilon_i)$ is termed as nonsystematic, idiosyncratic, or specific risk. This risk sis uncorrelated with the market and can be reduced by diversification.

#### 7.7. CAPM as Pricing Formula
a. Pricing form of the CAPM
The price $P$ of an asset with payoff $Q$ is
$$ P = \frac{\bar Q}{1 + r_f + \beta(\bar r_{M} - r_f)},$$
where $\beta$ is the beta of the asset. <br/>
b. Linearity of Pricing and the Certainty Equivalent Form
$$ P_1 + P_2 = \frac{\bar Q_1 + Q_2}{1 + r_f + \beta_{1+2}(\bar r_{M} - r_f)}, $$
where $\beta_{1+2}$ is the beta of a new asset which is the sum of assets 1 and 2.<br/>
c. Certainty equivalent pricing formula
The price of $P$ of an asset with payoff $Q$ is
$$ P = \frac{1}{1+r_f}[\bar Q - \frac{cov(Q,r_M)(\bar{r_M}-r_f)}{\sigma_M^2}]$$

for i = 1,2,...,m. In this equation, the $a_i$'s(intercepts) and $b_i$'s(factor loading) are fixed constants. The $e_i$'s are random quantities which represent errors.
The standdard parameters for mean-variance analysis can be determined this model.
$$ \bar r_i = a_i + b_i \bar f$$
$$ \sigma_i^2 = b_i^2 \sigma_f^2 + \sigma_{e_i}^2$$
$$ \sigma_{ij} = b_ib_j\sigma_f^2$$

$$ b = \sigma_{i=1}^n w_ib_i$$

Under the assumptions that $E(e_i)=0,\,E[(f-\bar f)e_i]=0, and E(e_ie_j)=0$ for all $i\neq j$, it is clear that $E(e)=0, E[(f-\bar f)e]=0$.e and f are uncorrelated. The variance of $e$ is

#### 8.4. Arbitrage Pricing Model
a. Simple APT <br/>
Suppose that there are n assets whose rates of return are governed by $m<n$ factors according to the equation
$$ r_i = a_i + \Sigma_{i=1}^{m}b_{ij}f_j$$
$for \, i = 1,2,...n$. Then there are constants $\lambda_0,\lambda_1,...,\lambda_m$ such that
$$ \bar r_i = \lambda_0 + \Sigma_{j=1}^m b_{ij}\lambda_j$$
$for \, i = 1,2,...n$. <br/>
b. Well-Diversified Portfolios <br/>
Let us form a portfolio using the weights $w_1, w_2, ... , w_n$ with $\Sigma_{i=1}^n w_i =1$. The return of the portfolio is
$$ r = a + \Sigma_{j=1}^{m}b_jf_j + e$$
where
$$ a= \Sigma_{i=1}^n w_ia_i$$
$$ b_j = \Sigma_{i=1}^n w_i b_{ij}$$
$$ \sigma_e^2 = \Sigma_{i=1}^n w_i^2\sigma_{e_i}^2$$
Suppose that for each i there holds $\sigma_{e_j}^2 \leq S^2$ for some constant S. Suppose also that the portfolio is well diversified in the sense that for each $i$ there holds $w_i \leq \frac{W}{n}$ for some constant $W \approx 1$.
$$ \sigma_e^2 \leq \frac{1}{n}\Sigma_{i=1}^n W^2S^2 \leq \frac{1}{n}W^2S^2$$
Let $n \to \infty$. While doing this we assume that the bound $\sigma_{e_i}^2 \leq S^2$ remains vaild for all i. Also for each n, we select a portfolio that is well diversified. As $n \to \infty$, we see that the $\sigma_e^2 \to 0$.In other words, the error term associated with a well-diversified protfolio of an infinite number of assets has as a variance of zero. <br/>
c.Relationship between CAPM and APT <br/>
The result of APT is that the coefficients of the underlying factor model must satisfy a linear relation. In the special case where the underlying factor model has the single factor equal to the excess return on the market portfolio, the CAPM theory states that $\alpha = 0$. This is a special case of APT, which states that the constant $a$ in the expression for the return of an asset is a linear combination of the factor loadings of that asset.

#### 8.5. Projection Pricing with Factors
Suppose there are $n$ such assets that span a significant submarket described by factors. In a manner similar to a standard framework, each asset i in that submarket has payoff
$$ y_i = a_i + \Sigma_{j=1}^m b_{i,j}f_j = \epsilon_i,$$
where $y_i$ is the payoff of the $i$th asset.
A new asset has payoff $y$ given by
$$ y = a + \Sigma_{j=1}^m b_jf_j + \epsilon$$
where, consistent with the structure of the n given assets, we assume that $\epsilon$ is independent of all other uncertainties.
To price this asset by projection we use the scaled minimum-norm portfolio of marketed assets having unit price. This scaled portfolio, being a linear combination of the original assets, is of the form
$$ y^q = a^q + \Sigma_{j=1}^m b_j^qf_j + \epsilon^q$$
The projection price of y is then found by taking the expectation of the product, $P_y=E[y^qy]$
This works ouyt to be
$$ P_y = \theta_0a+\sigma_{j=1}^m\theta_jb_j,$$
where 
$$ \theta_0= a^q + \Sigma_{j=1}^mb_j^q \bar f_j$$
$$ \theta_j = a^q \bar f_j + \Sigma_{i=1}^m b_i^q \bar {f_if_j}$$

 Let $w=[w_1,…,w_N ]^T$ be the weights of the risky assets in the portfolio.
Using the Lagrange Multiplier Methods, Find the mean return , volatility and optimal weight of the Global Minimum Variance(GMV) Portfolio using the Lagrange multiplier methods

#### sol)

$$ \displaystyle\min_w w^{T}\Sigma w \;\;s.t\;\; w^{T}1_N = 1 $$
Using the Lagrange multiplier methods,
$$ \mathcal{L} = w^{T}\Sigma w + \lambda (1- w^{T}1_N)$$
$ F.O.C \Rightarrow \frac{\partial \mathcal{L}}{\partial w} = 2\Sigma w - \lambda 1_N = 0 $ <br/>
$ 2\Sigma w = \lambda 1_N $
$$ w^* = \frac{\lambda}{2}\Sigma^{-1}1_N$$
Inserting $w^*$ into the two constraints,
$$ w^{T}1_N = 1$$
$$ \frac{\lambda}{2}1_N^T\Sigma^{-1}1_N =1$$ 
$$ \frac{\lambda}{2} = (1_N^T\Sigma^{-1}1_N)^{-1}$$
Therefore, the optimal weights of the GMV portfolio is
$$ w_{GMV} = \frac{\Sigma^{-1}1_N}{1_N^T\Sigma^{-1}1_N}$$
The mean return is
$$ \mu_{GMV} = w_{GMV}^T\mu = \frac{1_N^T\Sigma^{-1}\mu}{1_N^T\Sigma^{-1}1_N} $$
The variance of the portfolio is
$$ \sigma_{GMV}^2 = w_{GMV}^T\Sigma^{-1}w_{GMV} = (1_N^T\Sigma^{-1}1_N)^{-1} $$
And the volatility of the GMV portfolio is
$$  \sigma_{GMV} = (1_N^T\Sigma^{-1}1_N)^{-\frac{1}{2}}$$

## Problem 3. 
Explain that Tangency Portfolio becomes Market Portfolio under the assumption of CAPM, and show that CAL of Tangency Portfolio becomes CML (Capital Market Line).

#### sol)
A.Tangency Portfolio becomes Market Portfolio under the assumption of CAPM <br/>
$\to$ From the one-fund theorem, we know that everyone will form a portfolio that is mix of the risk-free asset and single risky one fund.Furthermore, since everyone use the same means, variances and covariances, everyone will use the same risky fund. Since the CAPM have the assumption which are perfect capital market, in order for supply to equal demand, the tangency protfolio must be the market portfolio.

B. show that CAL of Tangency Portfolio becomes CML (Capital Market Line) <br/>
Suppose that the investor has portfolio $p$ consisting of the market portfolio and risk-free asset. <br/>
The return of portfolio is 
$$ r_p = wr_f + (1-w)r_m$$
The expected return of portfolio and stanadard deviation are
$$ E(r_p) = wr_f + (1-w)E(r_m)$$
$$ \sigma_p = (1-w)\sigma_m$$
Combining both equations, we can derive the Capital Market Line(CML).
$$ E(r_p) = r_f + \frac{E(r_m) - r_f}{\sigma_m}\sigma_p$$
Since the $E(r_p)=\mu_p$ and $\frac{E(r_m) - r_f}{\sigma_m}$ is Sharpe Ratio, the CAL of Tangency Portfolio becomes CML .
$$ E(r_p) = r_f + \frac{E(r_m) - r_f}{\sigma_m}\sigma_p = SR \sigma_p + r_f = \mu_p $$

##### 1) Follow the proof on page 184 of the textbook to induce the formula of CAPM.

#### sol)
For any $\alpha$ consider the portfolio consisting of a portion $\alpha$ invested in asset i and a portion $1-\alpha$ invested in the market portfolio $M$ (We allow $\alpha <0$). The expected rate of return of this portfoio is
$$ \bar r_\alpha = \alpha \bar r_i + (1-\alpha) \bar r_M$$
and the standard deviation of the rate of return is 
$$ \sigma_\alpha = [\alpha^2\sigma_i^2 + 2\alpha(1-\alpha)\sigma_{iM}+(1-\alpha)^2\sigma_M^2]^{\frac{1}{2}}$$
If $\alpha = 0$, the protfolio is corresponds to the market portfolio. The tangency condition can be translated into the condition that the slope of the curve is equal to the slope of the capital market line at the point of the makret portfolio.
$$ \frac{d\bar r_\alpha}{d \alpha} = \bar r_i - \bar r_M$$
$$ \frac{d\bar \sigma_\alpha}{d \alpha} = \frac{\alpha \sigma_i^2 +(1-2\alpha)\sigma_{iM}+(\alpha -1)\sigma_M^2}{\sigma_\alpha}$$
When $\alpha = 0$,
$$ \frac{d\bar \sigma_\alpha}{d \alpha} |_{\alpha = 0} = \frac{\sigma_{iM}-\sigma_M^2}{\sigma_M} $$
Using the relation,
$$ \frac{d\bar r_\alpha}{d\bar \sigma_\alpha} = \frac{d\bar r_\alpha}{d \alpha}\frac{d\alpha}{d\bar \sigma_\alpha}$$
to obtain
$$ \frac{d\bar r_\alpha}{d\bar \sigma_\alpha}|_{\alpha=0} = \frac{(\bar r_i - \bar r_M)\sigma_M}{\sigma_{iM}-\sigma_M^2}$$
This slope must equal the slope of the CML.
$$ \frac{(\bar r_i - \bar r_M)\sigma_M}{\sigma_{iM}-\sigma_M^2} = \frac{\bar r_M - r_f}{\sigma_M} $$
Solving for $\bar r_i$, obtaining the formula of CAPM.
$$ \bar r_i = r_f + (\frac{\bar r_M - r_f}{\sigma_M^2})\sigma_{iM} = r_f + \beta(\bar r_M - r_f)$$
